# Introducción a BigQuery y Snowflake



## Introducción

En los últimos años, las organizaciones han incrementado significativamente la cantidad de datos que generan y almacenan. Este crecimiento ha impulsado la adopción de plataformas analíticas en la nube capaces de procesar grandes volúmenes de información de forma rápida, escalable y rentable.

Dentro de este contexto, BigQuery y Snowflake se han convertido en dos de las soluciones más utilizadas por empresas de todos los tamaños para construir plataformas modernas de análisis de datos, Business Intelligence (BI), Machine Learning e Ingeniería de Datos.

Ambas herramientas permiten almacenar y analizar terabytes o incluso petabytes de información sin la necesidad de administrar servidores físicos o infraestructura compleja.

## Objetivos de Aprendizaje

Al finalizar este tema, el estudiante será capaz de:

* Comprender qué son BigQuery y Snowflake.
* Identificar las principales diferencias entre ambas plataformas.
* Reconocer los escenarios en los que cada solución es más adecuada.
* Entender su papel dentro de una arquitectura moderna de datos.

## ¿Qué es BigQuery?

BigQuery es el servicio de Data Warehouse Serverless de Google Cloud diseñado para almacenar y analizar grandes volúmenes de datos mediante consultas SQL de alta velocidad.

Una de sus principales características es que Google administra completamente la infraestructura, por lo que los usuarios únicamente se preocupan por almacenar datos y ejecutar consultas.

### Características principales

* Totalmente Serverless.
* Escalabilidad automática.
* Procesamiento distribuido.
* Integración nativa con Google Cloud.
* Compatible con SQL estándar.
* Capacidad para procesar petabytes de información.

### Casos de uso comunes

* Dashboards empresariales.
* Análisis de ventas.
* Analítica de marketing.
* Machine Learning.
* Procesamiento de logs.
* Análisis de comportamiento de usuarios.

## ¿Qué es Snowflake?

Snowflake es una plataforma de datos en la nube que combina almacenamiento, procesamiento y compartición de datos en una única solución altamente escalable.

A diferencia de BigQuery, Snowflake permite administrar de forma independiente los recursos de almacenamiento y procesamiento mediante componentes llamados Virtual Warehouses.

### Características principales

* Arquitectura multicapa.
* Escalabilidad independiente del almacenamiento y cómputo.
* Compartición segura de datos.
* Compatibilidad con múltiples proveedores cloud.
* Alto rendimiento para cargas analíticas.

### Casos de uso comunes

* Data Warehouses empresariales.
* Plataformas de datos corporativas.
* Data Sharing entre organizaciones.
* Data Lakes modernos.
* Ambientes multiusuario de alta concurrencia.

## BigQuery vs Snowflake: Visión General

| Característica        | BigQuery                      | Snowflake                    |
| --------------------- | ----------------------------- | ---------------------------- |
| Proveedor             | Google Cloud                  | Multi Cloud                  |
| Infraestructura       | Serverless                    | Virtual Warehouses           |
| Escalabilidad         | Automática                    | Configurable                 |
| Modelo de pago        | Por consulta y almacenamiento | Por almacenamiento y cómputo |
| Administración        | Mínima                        | Moderada                     |
| Data Sharing          | Limitado                      | Muy avanzado                 |
| Integración principal | Ecosistema Google             | AWS, Azure y Google Cloud    |

## Ejemplo Empresarial

Supongamos que una empresa de comercio electrónico genera diariamente millones de registros de ventas.

### Escenario con BigQuery

La empresa almacena los datos en Google Cloud Storage y utiliza BigQuery para:

* Analizar ventas diarias.
* Construir dashboards en Looker Studio.
* Crear modelos predictivos con Vertex AI.

La infraestructura se escala automáticamente sin intervención del equipo técnico.

### Escenario con Snowflake

La misma empresa trabaja con varias filiales ubicadas en distintos países.

Snowflake permite:

* Compartir datos entre departamentos.
* Crear entornos independientes para analistas.
* Escalar recursos de procesamiento según la demanda.
* Integrarse simultáneamente con AWS y Azure.

## ¿Cuándo utilizar BigQuery?

BigQuery suele ser una excelente opción cuando:

* La organización trabaja principalmente con Google Cloud.
* Se desea minimizar la administración de infraestructura.
* Existen cargas analíticas variables.
* Se busca una implementación rápida.

## ¿Cuándo utilizar Snowflake?

Snowflake suele ser una excelente opción cuando:

* Se requiere alta concurrencia de usuarios.
* Existen múltiples equipos trabajando simultáneamente.
* Se necesita compartir datos con terceros.
* La empresa opera en ambientes Multi Cloud.

## Rol dentro de una Arquitectura de Datos Moderna

Actualmente BigQuery y Snowflake suelen ubicarse en la capa analítica de la arquitectura.

Flujo típico:

Datos Fuente → Ingesta → Data Lake → BigQuery/Snowflake → BI / Machine Learning

Ejemplos de fuentes:

* ERP
* CRM
* APIs
* Bases de datos transaccionales
* Archivos CSV
* Aplicaciones móviles
* Sensores IoT

Posteriormente los datos son consumidos por herramientas como:

* Power BI
* Tableau
* Looker
* Python
* dbt
* Airflow

## Ventajas de las Plataformas Cloud Modernas

Las soluciones modernas como BigQuery y Snowflake ofrecen beneficios importantes frente a los Data Warehouses tradicionales:

* Escalabilidad prácticamente ilimitada.
* Menor costo de infraestructura.
* Alta disponibilidad.
* Actualizaciones automáticas.
* Integración con herramientas modernas de análisis.
* Implementación más rápida.





# Arquitectura de BigQuery



## Introducción

Uno de los principales factores que han convertido a BigQuery en una de las plataformas analíticas más utilizadas es su arquitectura Serverless. A diferencia de las bases de datos tradicionales, donde los administradores deben gestionar servidores, almacenamiento y recursos de procesamiento, BigQuery abstrae completamente esta complejidad.

Esto permite que los equipos de datos se concentren en analizar información y desarrollar soluciones sin preocuparse por la infraestructura subyacente.

## Objetivos de Aprendizaje

Al finalizar este tema, el estudiante será capaz de:

* Comprender la arquitectura interna de BigQuery.
* Identificar los componentes principales de la plataforma.
* Entender cómo se ejecutan las consultas.
* Reconocer el papel de los Slots dentro del procesamiento distribuido.

## Arquitectura General de BigQuery

BigQuery está diseñado bajo una arquitectura desacoplada donde el almacenamiento y el procesamiento funcionan de manera independiente.

Esta separación permite escalar ambos componentes sin afectar el rendimiento general de la plataforma.

### Componentes Principales

* Storage Layer
* Compute Layer
* Query Engine
* Slots
* Metadata Services

## Storage Layer

Es la capa encargada de almacenar los datos.

BigQuery guarda la información utilizando un formato columnar altamente optimizado para consultas analíticas.

### Ventajas del almacenamiento columnar

* Lectura más rápida.
* Menor consumo de recursos.
* Compresión eficiente.
* Mejor rendimiento para agregaciones.

### Ejemplo

Supongamos una tabla de ventas:

| id_venta | fecha      | producto | cliente | monto |
| -------- | ---------- | -------- | ------- | ----- |
| 1        | 2025-01-01 | Laptop   | Juan    | 5000  |
| 2        | 2025-01-01 | Mouse    | María   | 100   |

Si una consulta solicita únicamente la columna monto:

```sql
SELECT monto
FROM ventas;
```

BigQuery leerá únicamente esa columna, reduciendo significativamente la cantidad de datos procesados.

## Compute Layer

Es la capa encargada de ejecutar las consultas.

Cuando un usuario envía una consulta SQL:

```sql
SELECT producto,
       SUM(monto)
FROM ventas
GROUP BY producto;
```

BigQuery distribuye automáticamente el trabajo entre múltiples nodos de procesamiento.

El usuario no necesita configurar servidores ni clústeres.

## Arquitectura Serverless

BigQuery es considerado un servicio Serverless porque:

* No requiere aprovisionar servidores.
* No requiere configurar clústeres.
* No requiere administrar capacidad.
* Escala automáticamente.

Desde la perspectiva del usuario únicamente existen:

* Datos almacenados.
* Consultas SQL.

Toda la infraestructura es gestionada por Google.

## Flujo de Ejecución de una Consulta

Cuando una consulta es enviada, ocurre el siguiente proceso:

### Paso 1

El usuario envía la consulta SQL.

```sql
SELECT *
FROM ventas;
```

### Paso 2

BigQuery analiza la sintaxis.

### Paso 3

El optimizador genera un plan de ejecución.

### Paso 4

El trabajo es distribuido entre múltiples Slots.

### Paso 5

Los resultados son consolidados.

### Paso 6

Los resultados son devueltos al usuario.

## ¿Qué son los Slots?

Los Slots son la unidad básica de procesamiento dentro de BigQuery.

Un Slot representa una combinación de:

* CPU
* Memoria
* Recursos de red

Cada consulta consume uno o varios Slots dependiendo de:

* Complejidad de la consulta.
* Volumen de datos.
* Operaciones realizadas.

## Ejemplo Conceptual

Consulta simple:

```sql
SELECT *
FROM clientes
LIMIT 100;
```

Puede utilizar pocos Slots.

Consulta compleja:

```sql
SELECT c.pais,
       SUM(v.monto)
FROM ventas v
JOIN clientes c
ON v.id_cliente = c.id_cliente
GROUP BY c.pais;
```

Utilizará una mayor cantidad de Slots debido al procesamiento adicional.

## Procesamiento Distribuido

BigQuery divide automáticamente grandes conjuntos de datos en múltiples fragmentos.

Por ejemplo:

Tabla de ventas:

| Registros     |
| ------------- |
| 5,000,000,000 |

En lugar de procesar toda la tabla en un único servidor:

* Nodo 1 procesa una parte.
* Nodo 2 procesa otra parte.
* Nodo 3 procesa otra parte.

Los resultados se combinan posteriormente.

Este enfoque permite analizar terabytes o petabytes de información en segundos o minutos.

## Almacenamiento y Procesamiento Desacoplados

Una característica clave de BigQuery es que el almacenamiento y el procesamiento son independientes.

### Beneficios

* Escalabilidad flexible.
* Menor costo operativo.
* Mayor disponibilidad.
* Mejor aprovechamiento de recursos.

| Componente | Función           |
| ---------- | ----------------- |
| Storage    | Guarda los datos  |
| Compute    | Ejecuta consultas |

## Optimizador de Consultas

Antes de ejecutar una consulta, BigQuery genera automáticamente el plan más eficiente posible.

Algunas optimizaciones incluyen:

* Eliminación de lecturas innecesarias.
* Reordenamiento de operaciones.
* Uso eficiente de particiones.
* Paralelización automática.

Por esta razón dos consultas que producen el mismo resultado pueden tener costos muy diferentes.

## Ejemplo de Buena Práctica

Consulta poco eficiente:

```sql
SELECT *
FROM ventas;
```

Consulta optimizada:

```sql
SELECT fecha,
       producto,
       monto
FROM ventas;
```

La segunda consulta procesa menos datos y reduce costos.

## Caso de Uso Empresarial

Una cadena de supermercados genera diariamente:

* 20 millones de ventas.
* 5 millones de movimientos de inventario.
* 1 millón de registros logísticos.

BigQuery permite:

* Consolidar toda la información.
* Ejecutar reportes en tiempo real.
* Analizar tendencias de ventas.
* Alimentar dashboards ejecutivos.

Todo esto sin administrar infraestructura física ni servidores dedicados.

## Buenas Prácticas

* Consultar únicamente las columnas necesarias.
* Utilizar particionamiento cuando sea posible.
* Evitar SELECT * en ambientes productivos.
* Monitorear el volumen de datos procesados.
* Diseñar consultas eficientes.

## Resumen

La arquitectura de BigQuery se basa en una separación clara entre almacenamiento y procesamiento, permitiendo una escalabilidad prácticamente ilimitada. Su modelo Serverless elimina la necesidad de administrar infraestructura y utiliza procesamiento distribuido mediante Slots para ejecutar consultas de forma eficiente sobre grandes volúmenes de datos. Estas características convierten a BigQuery en una de las plataformas analíticas más potentes dentro del ecosistema moderno de Ingeniería de Datos.


# Arquitectura de Snowflake



## Introducción

Snowflake fue diseñado específicamente para resolver algunas de las limitaciones que presentaban los Data Warehouses tradicionales. Su principal innovación consiste en separar completamente el almacenamiento, el procesamiento y los servicios de administración.

Esta arquitectura permite que múltiples equipos trabajen simultáneamente sobre los mismos datos sin afectar el rendimiento entre ellos.

## Objetivos de Aprendizaje

Al finalizar este tema, el estudiante será capaz de:

* Comprender la arquitectura de Snowflake.
* Identificar las tres capas principales de la plataforma.
* Entender el funcionamiento de los Virtual Warehouses.
* Reconocer las ventajas de la separación entre almacenamiento y cómputo.

## Arquitectura General de Snowflake

Snowflake está compuesto por tres capas principales:

1. Database Storage Layer
2. Compute Layer
3. Cloud Services Layer

Cada capa cumple una función específica dentro de la plataforma.

## Database Storage Layer

Es la capa encargada de almacenar todos los datos.

Cuando los datos son cargados en Snowflake:

* Son comprimidos automáticamente.
* Son organizados en formato columnar.
* Son distribuidos en Micro-Partitions.
* Son almacenados en el proveedor cloud correspondiente.

El usuario no necesita administrar archivos ni estructuras físicas.

### Características principales

* Compresión automática.
* Almacenamiento columnar.
* Escalabilidad automática.
* Alta disponibilidad.

## ¿Qué son las Micro-Partitions?

Snowflake divide automáticamente las tablas en pequeños bloques llamados Micro-Partitions.

Cada Micro-Partition contiene:

* Datos.
* Estadísticas.
* Metadatos.

Estas estadísticas permiten que Snowflake identifique rápidamente qué información necesita leer y cuál puede ignorar.

### Ejemplo

Tabla de ventas:

| Fecha      | Producto | Monto |
| ---------- | -------- | ----- |
| 2025-01-01 | Laptop   | 5000  |
| 2025-01-02 | Mouse    | 100   |
| 2025-01-03 | Monitor  | 2000  |

Consulta:

```sql
SELECT *
FROM ventas
WHERE fecha = '2025-01-03';
```

Snowflake leerá únicamente las Micro-Partitions relevantes para esa fecha.

Este proceso se conoce como Data Pruning.

## Compute Layer

Es la capa responsable de ejecutar consultas SQL.

A diferencia de BigQuery, Snowflake utiliza recursos de procesamiento llamados Virtual Warehouses.

Cada Virtual Warehouse funciona como un clúster de procesamiento independiente.

## ¿Qué es un Virtual Warehouse?

Un Virtual Warehouse es un conjunto de recursos computacionales utilizados para ejecutar:

* Consultas SQL.
* Procesos ETL.
* Cargas de datos.
* Transformaciones.

Los Virtual Warehouses pueden crearse, modificarse o eliminarse según las necesidades del negocio.

### Ejemplo

Una empresa puede tener:

| Warehouse       | Uso                |
| --------------- | ------------------ |
| ETL_WH          | Procesos de carga  |
| BI_WH           | Dashboards         |
| DATA_SCIENCE_WH | Modelos analíticos |

Cada equipo utiliza recursos independientes sin afectar a los demás.

## Escalabilidad de los Virtual Warehouses

Snowflake permite aumentar o reducir la capacidad de procesamiento fácilmente.

Tamaños comunes:

* X-Small
* Small
* Medium
* Large
* X-Large

### Ejemplo

Proceso diario:

```text
ETL nocturno → Large
Horario normal → Small
Fin de mes → X-Large
```

Esto permite optimizar costos y rendimiento.

## Multi-Cluster Warehouses

Cuando existen muchos usuarios concurrentes, Snowflake puede iniciar múltiples clústeres automáticamente.

### Beneficios

* Evita cuellos de botella.
* Mantiene tiempos de respuesta estables.
* Incrementa la concurrencia.

Ejemplo:

50 usuarios ejecutando reportes simultáneamente.

Snowflake distribuye la carga entre varios clústeres para mantener el rendimiento.

## Cloud Services Layer

Es la capa encargada de coordinar todo el funcionamiento de la plataforma.

Entre sus responsabilidades se encuentran:

* Autenticación.
* Seguridad.
* Gestión de metadatos.
* Optimización de consultas.
* Control de acceso.
* Administración de sesiones.

Esta capa es invisible para el usuario, pero es fundamental para el funcionamiento de Snowflake.

## Flujo de Ejecución de una Consulta

Consulta:

```sql
SELECT producto,
       SUM(monto)
FROM ventas
GROUP BY producto;
```

Proceso interno:

### Paso 1

El usuario envía la consulta.

### Paso 2

Cloud Services analiza la sintaxis.

### Paso 3

Se genera un plan de ejecución optimizado.

### Paso 4

El Virtual Warehouse ejecuta la consulta.

### Paso 5

Se leen únicamente las Micro-Partitions necesarias.

### Paso 6

Los resultados son enviados al usuario.

## Separación entre Storage y Compute

Una de las características más importantes de Snowflake es que el almacenamiento y el procesamiento son completamente independientes.

| Componente     | Función                  |
| -------------- | ------------------------ |
| Storage Layer  | Almacena datos           |
| Compute Layer  | Procesa consultas        |
| Cloud Services | Administra la plataforma |

### Beneficios

* Escalamiento independiente.
* Mayor flexibilidad.
* Optimización de costos.
* Mejor rendimiento para múltiples usuarios.

## Arquitectura Multi Cloud

Snowflake puede ejecutarse sobre:

* Amazon Web Services
* Microsoft Azure
* Google Cloud

Esto permite a las organizaciones seleccionar el proveedor que mejor se adapte a sus necesidades.

## Caso de Uso Empresarial

Una empresa multinacional posee:

* Equipo de Ingeniería de Datos.
* Equipo de BI.
* Equipo de Ciencia de Datos.

Cada equipo ejecuta consultas constantemente.

Con Snowflake se crean tres Virtual Warehouses independientes:

| Equipo              | Warehouse    |
| ------------------- | ------------ |
| Ingeniería de Datos | ETL_WH       |
| BI                  | REPORTING_WH |
| Ciencia de Datos    | ML_WH        |

De esta forma, una consulta pesada de Ciencia de Datos no afecta los dashboards utilizados por la dirección ejecutiva.

## Buenas Prácticas

* Crear Virtual Warehouses según el tipo de carga.
* Utilizar Auto Suspend para reducir costos.
* Utilizar Auto Resume para mejorar la experiencia de usuario.
* Monitorear el consumo de créditos.
* Evitar Warehouses sobredimensionados.

## Resumen

La arquitectura de Snowflake se basa en la separación de almacenamiento, procesamiento y servicios administrativos. Su uso de Micro-Partitions permite optimizar la lectura de datos, mientras que los Virtual Warehouses ofrecen recursos de cómputo independientes para distintos equipos y procesos. Esta arquitectura proporciona una alta escalabilidad, excelente concurrencia y una gestión eficiente de costos, convirtiendo a Snowflake en una de las plataformas de datos más utilizadas en entornos empresariales modernos.


# BigQuery vs Snowflake



## Introducción

BigQuery y Snowflake son dos de las plataformas analíticas más utilizadas en la actualidad. Ambas permiten almacenar y procesar grandes volúmenes de información, integrarse con herramientas de Business Intelligence y soportar arquitecturas modernas de datos.

Sin embargo, cada una fue diseñada con enfoques diferentes, por lo que elegir la solución adecuada dependerá de las necesidades técnicas, operativas y financieras de cada organización.

## Objetivos de Aprendizaje

Al finalizar este tema, el estudiante será capaz de:

* Comparar BigQuery y Snowflake.
* Identificar las ventajas y desventajas de cada plataforma.
* Comprender los modelos de costos.
* Seleccionar la plataforma adecuada según distintos escenarios empresariales.

## Comparación General

| Característica                    | BigQuery         | Snowflake             |
| --------------------------------- | ---------------- | --------------------- |
| Tipo de servicio                  | Serverless       | Plataforma gestionada |
| Proveedor                         | Google Cloud     | Multi Cloud           |
| Administración de infraestructura | No requerida     | Mínima                |
| Escalabilidad                     | Automática       | Configurable          |
| Procesamiento                     | Slots            | Virtual Warehouses    |
| Modelo de pago principal          | Datos procesados | Créditos consumidos   |
| Compartición de datos             | Básica           | Muy avanzada          |
| Concurrencia                      | Muy buena        | Excelente             |
| Multi Cloud                       | No               | Sí                    |

## Diferencias Arquitectónicas

### BigQuery

BigQuery fue diseñado bajo un modelo completamente Serverless.

El usuario:

* Carga datos.
* Ejecuta consultas.
* Consume resultados.

Google administra toda la infraestructura.

### Snowflake

Snowflake permite controlar el cómputo mediante Virtual Warehouses.

El usuario puede:

* Crear distintos clústeres.
* Escalar recursos.
* Asignar recursos a equipos específicos.

Esto proporciona mayor flexibilidad operativa.

## Comparación de Escalabilidad

### BigQuery

La escalabilidad es automática.

Ejemplo:

Si una consulta pasa de procesar 10 GB a 10 TB, BigQuery ajusta automáticamente los recursos necesarios.

Ventajas:

* Menor administración.
* Configuración mínima.
* Fácil adopción.

### Snowflake

La escalabilidad es configurable.

Ejemplo:

Un Warehouse puede cambiar de:

```text
Small → Medium → Large → X-Large
```

Ventajas:

* Mayor control.
* Optimización de costos.
* Ajuste fino del rendimiento.

## Comparación de Costos

### BigQuery

Generalmente cobra por:

* Almacenamiento.
* Datos procesados por consulta.

Ejemplo:

Consulta:

```sql
SELECT *
FROM ventas;
```

Si la tabla contiene 500 GB, se cobrará según el volumen procesado.

### Snowflake

Generalmente cobra por:

* Almacenamiento.
* Tiempo de uso de los Virtual Warehouses.

Ejemplo:

Warehouse activo:

```text
Small Warehouse
Activo durante 2 horas
```

El costo dependerá del tiempo de ejecución y tamaño del Warehouse.

## ¿Cuál es más económico?

Depende del patrón de uso.

### BigQuery suele ser más conveniente cuando

* Las consultas son esporádicas.
* Existen picos de uso impredecibles.
* Se desea minimizar la administración.

### Snowflake suele ser más conveniente cuando

* Existen cargas de trabajo constantes.
* Se requiere alta concurrencia.
* Hay múltiples equipos utilizando la plataforma simultáneamente.

## Comparación de Rendimiento

### BigQuery

Destaca en:

* Consultas masivas.
* Procesamiento distribuido.
* Análisis de grandes volúmenes de datos.

Ejemplos:

* Logs.
* Eventos web.
* IoT.
* Datos históricos.

### Snowflake

Destaca en:

* Ambientes multiusuario.
* Procesamiento concurrente.
* Data Sharing.
* Entornos corporativos complejos.

Ejemplos:

* Finanzas.
* Bancos.
* Corporaciones multinacionales.

### BigQuery

Integración nativa con:

### Snowflake

Integración con:

* [Amazon Web Services](https://aws.amazon.com?utm_source=chatgpt.com)

* [Google Cloud](https://cloud.google.com?utm_source=chatgpt.com)
* Tableau
* Power BI
* Airflow
* dbt

## Comparación de Compartición de Datos

### BigQuery

Permite compartir:

* Datasets.
* Tablas.
* Vistas.

Adecuado para la mayoría de escenarios internos.

### Snowflake

Incluye funcionalidades avanzadas de Data Sharing.

Permite:

* Compartir datos entre empresas.
* Evitar copias físicas.
* Mantener una única versión de los datos.

Esta es una de las características más valoradas de Snowflake.

## Comparación de Curva de Aprendizaje

| Aspecto               | BigQuery   | Snowflake |
| --------------------- | ---------- | --------- |
| Configuración inicial | Muy simple | Simple    |
| Administración        | Muy baja   | Moderada  |
| Tiempo de adopción    | Rápido     | Rápido    |
| Complejidad operativa | Baja       | Media     |

## Casos de Uso Recomendados

### Utilizar BigQuery cuando

* La empresa trabaja principalmente con Google Cloud.
* Se requiere una solución Serverless.
* Los equipos son pequeños o medianos.
* Se prioriza la simplicidad operativa.

### Utilizar Snowflake cuando

* Existen múltiples áreas analíticas.
* Se requiere alta concurrencia.
* Se necesita compartir datos con terceros.
* Se trabaja en ambientes Multi Cloud.

## Escenario Empresarial Comparativo

Una empresa de comercio electrónico procesa 5 TB diarios de información.

### Solución BigQuery

* Datos almacenados en Google Cloud.
* Dashboards en Looker Studio.
* Procesamiento automático.
* Administración mínima.

### Solución Snowflake

* Datos distribuidos entre AWS y Azure.
* Equipos de BI, Data Science y Finanzas.
* Warehouses independientes.
* Alta concurrencia.

Ambas soluciones son técnicamente válidas, pero la elección dependerá de la estrategia tecnológica de la organización.

## Ventajas y Desventajas

### BigQuery

#### Ventajas

* Serverless.
* Escalabilidad automática.
* Administración mínima.
* Excelente integración con Google Cloud.

#### Desventajas

* Menor control sobre el cómputo.
* Dependencia del ecosistema Google.

### Snowflake

#### Ventajas

* Multi Cloud.
* Excelente concurrencia.
* Data Sharing avanzado.
* Gran flexibilidad operativa.

#### Desventajas

* Mayor administración.
* Gestión de Warehouses y créditos.

## Resumen

BigQuery y Snowflake son plataformas líderes en el mercado de datos analíticos. BigQuery destaca por su simplicidad, modelo Serverless y estrecha integración con Google Cloud, mientras que Snowflake sobresale por su arquitectura multicapa, capacidad Multi Cloud y administración independiente del procesamiento mediante Virtual Warehouses.

No existe una solución universalmente mejor. La elección dependerá de factores como la infraestructura existente, los patrones de consumo, la concurrencia requerida, el presupuesto y los objetivos estratégicos de la organización.


# Objetos Principales en BigQuery

## Introducción

Antes de comenzar a cargar datos o ejecutar consultas, es importante comprender cómo organiza BigQuery la información internamente.

BigQuery utiliza una estructura jerárquica simple que permite organizar proyectos, conjuntos de datos y tablas de manera ordenada.

Comprender estos objetos es fundamental para administrar correctamente ambientes de desarrollo, pruebas y producción.

## Objetivos de Aprendizaje

Al finalizar este tema, el estudiante será capaz de:

* Identificar los principales objetos de BigQuery.
* Comprender la jerarquía de almacenamiento.
* Crear tablas y vistas.
* Diferenciar entre tablas, vistas y vistas materializadas.

## Jerarquía de Objetos en BigQuery

La estructura básica de BigQuery es:

```text
Project
 └── Dataset
      ├── Table
      ├── View
      └── Materialized View
```

Cada nivel cumple una función específica.

## Project

El Project es el nivel más alto dentro de Google Cloud.

Todo recurso de BigQuery pertenece obligatoriamente a un Project.

Ejemplo:

```text
empresa-analytics
```

Dentro de un proyecto pueden existir múltiples datasets.

### Casos comunes

* Desarrollo
* QA
* Producción
* Sandbox

Ejemplo:

```text
empresa-dev
empresa-qa
empresa-prod
```

## Dataset

Un Dataset funciona como un contenedor lógico de tablas y vistas.

Es similar a un esquema (Schema) en otros motores de bases de datos.

Ejemplo:

```text
ventas
marketing
recursos_humanos
finanzas
```

Dentro de un Dataset pueden almacenarse cientos o miles de tablas.

## Creación de un Dataset

```sql
CREATE SCHEMA ventas;
```

Resultado:

| Dataset |
| ------- |
| ventas  |

## Table

Las tablas almacenan físicamente los datos.

Ejemplo:

```sql
CREATE TABLE ventas.clientes (
    id_cliente INT64,
    nombre STRING,
    correo STRING
);
```

Resultado esperado:

| id_cliente | nombre     | correo                                  |
| ---------- | ---------- | --------------------------------------- |
| 1          | Juan Pérez | [juan@email.com](mailto:juan@email.com) |
| 2          | Ana López  | [ana@email.com](mailto:ana@email.com)   |

## Inserción de Datos

```sql
INSERT INTO ventas.clientes
VALUES
(1,'Juan Pérez','juan@email.com'),
(2,'Ana López','ana@email.com');
```

## Consulta de Datos

```sql
SELECT *
FROM ventas.clientes;
```

Resultado:

| id_cliente | nombre     | correo                                  |
| ---------- | ---------- | --------------------------------------- |
| 1          | Juan Pérez | [juan@email.com](mailto:juan@email.com) |
| 2          | Ana López  | [ana@email.com](mailto:ana@email.com)   |

## View

Una View es una consulta almacenada.

No guarda datos físicamente.

Cada vez que se consulta, BigQuery ejecuta nuevamente la sentencia SQL.

Ejemplo:

```sql
CREATE VIEW ventas.v_clientes AS
SELECT
    id_cliente,
    nombre
FROM ventas.clientes;
```

Consulta:

```sql
SELECT *
FROM ventas.v_clientes;
```

Resultado:

| id_cliente | nombre     |
| ---------- | ---------- |
| 1          | Juan Pérez |
| 2          | Ana López  |

## Ventajas de las Views

* Simplifican consultas complejas.
* Mejoran la organización.
* Facilitan la seguridad.
* Evitan duplicar lógica SQL.

## Materialized View

Una Materialized View almacena físicamente los resultados de una consulta.

A diferencia de una View tradicional:

| Característica | View           | Materialized View |
| -------------- | -------------- | ----------------- |
| Guarda datos   | No             | Sí                |
| Actualización  | En tiempo real | Automática        |
| Rendimiento    | Normal         | Más rápido        |

## Ejemplo

```sql
CREATE MATERIALIZED VIEW ventas.mv_total_ventas
AS
SELECT
    producto,
    SUM(monto) AS total_ventas
FROM ventas.transacciones
GROUP BY producto;
```

Consulta:

```sql
SELECT *
FROM ventas.mv_total_ventas;
```

La respuesta será mucho más rápida porque parte del cálculo ya fue realizado previamente.

## Tabla Externa (External Table)

BigQuery permite consultar archivos sin cargarlos físicamente.

Ejemplo:

Archivo:

```text
ventas.csv
```

Ubicación:

```text
Google Cloud Storage
```

Consulta:

```sql
SELECT *
FROM ventas.tabla_externa;
```

Los datos permanecen en el almacenamiento externo.

### Ventajas

* Menor almacenamiento.
* Implementación rápida.
* Integración sencilla.

## Información de Metadatos

BigQuery proporciona información sobre los objetos mediante vistas de sistema.

Ejemplo:

```sql
SELECT *
FROM ventas.INFORMATION_SCHEMA.TABLES;
```

Resultado:

| table_name |
| ---------- |
| clientes   |
| productos  |
| ventas     |

## Nomenclatura Completa

En BigQuery una tabla suele identificarse mediante:

```text
project.dataset.table
```

Ejemplo:

```text
empresa-prod.ventas.clientes
```

Consulta:

```sql
SELECT *
FROM `empresa-prod.ventas.clientes`;
```

## Caso de Uso Empresarial

Una empresa posee el siguiente proyecto:

```text
empresa-prod
```

Datasets:

```text
ventas
marketing
finanzas
```

Tablas:

```text
ventas.transacciones
ventas.productos
marketing.campañas
finanzas.presupuestos
```

Views:

```text
ventas.v_resumen_mensual
```

Materialized Views:

```text
ventas.mv_top_productos
```

Esta organización permite administrar fácilmente grandes volúmenes de información.

## Buenas Prácticas

* Separar ambientes por proyecto.
* Organizar datasets por dominio de negocio.
* Utilizar nombres descriptivos.
* Crear vistas para reutilizar lógica SQL.
* Utilizar Materialized Views en consultas frecuentes.
* Evitar almacenar información duplicada.

## Resumen

BigQuery organiza sus recursos mediante una jerarquía compuesta por Projects, Datasets, Tables, Views y Materialized Views. Comprender estos objetos es esencial para diseñar soluciones analíticas escalables y fáciles de administrar. Una correcta organización mejora el rendimiento, facilita el mantenimiento y reduce la complejidad de los proyectos de Ingeniería de Datos.


# Objetos Principales en Snowflake

## Introducción

Snowflake utiliza una estructura jerárquica diferente a BigQuery. Aunque ambos permiten organizar información analítica, Snowflake incorpora algunos componentes adicionales relacionados con el procesamiento de datos y la administración de recursos computacionales.

Comprender estos objetos es fundamental para administrar correctamente una plataforma Snowflake en ambientes empresariales.

## Jerarquía General

La estructura básica de Snowflake es:

```text
Account
 └── Database
      └── Schema
           ├── Table
           ├── View
           ├── Stage
           └── File Format

Virtual Warehouse
```

A diferencia de BigQuery, los recursos de procesamiento (Warehouses) están separados de los objetos de almacenamiento.

## Database

Una Database es el contenedor principal de información.

Ejemplo:

```sql
CREATE DATABASE ventas_db;
```

Resultado:

| Database  |
| --------- |
| ventas_db |

Una organización puede tener varias bases de datos:

```text
ventas_db
marketing_db
finanzas_db
rrhh_db
```

## Schema

Los Schemas permiten organizar los objetos dentro de una Database.

Ejemplo:

```sql
CREATE SCHEMA ventas_db.comercial;
```

Resultado:

```text
ventas_db
 └── comercial
```

Es equivalente a los Datasets en BigQuery.

## Tables

Las tablas almacenan físicamente la información.

Ejemplo:

```sql
CREATE TABLE clientes (
    id_cliente INTEGER,
    nombre STRING,
    correo STRING
);
```

Inserción de datos:

```sql
INSERT INTO clientes
VALUES
(1,'Juan Pérez','juan@email.com'),
(2,'Ana López','ana@email.com');
```

Consulta:

```sql
SELECT *
FROM clientes;
```

Resultado:

| id_cliente | nombre     | correo                                  |
| ---------- | ---------- | --------------------------------------- |
| 1          | Juan Pérez | [juan@email.com](mailto:juan@email.com) |
| 2          | Ana López  | [ana@email.com](mailto:ana@email.com)   |

## Views

Las Views funcionan de manera similar a otros motores SQL.

Ejemplo:

```sql
CREATE VIEW v_clientes AS
SELECT
    id_cliente,
    nombre
FROM clientes;
```

Consulta:

```sql
SELECT *
FROM v_clientes;
```

Resultado:

| id_cliente | nombre     |
| ---------- | ---------- |
| 1          | Juan Pérez |
| 2          | Ana López  |

## Materialized Views

Las Materialized Views almacenan físicamente los resultados de una consulta.

Ejemplo:

```sql
CREATE MATERIALIZED VIEW mv_ventas_producto AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas
GROUP BY producto;
```

Ventajas:

* Consultas más rápidas.
* Menor tiempo de respuesta.
* Menor consumo de recursos en consultas repetitivas.

## Stages

Los Stages son una de las características más importantes de Snowflake.

Se utilizan para almacenar temporalmente archivos antes de cargarlos a tablas.

Ejemplo:

```sql
CREATE STAGE carga_clientes;
```

Archivos típicos:

```text
clientes.csv
ventas.parquet
productos.json
```

## Tipos de Stages

### Internal Stage

Los archivos se almacenan dentro de Snowflake.

```sql
CREATE STAGE stage_interno;
```

### External Stage

Los archivos permanecen en almacenamiento externo.

Ejemplos:

* AWS S3
* Azure Blob Storage
* Google Cloud Storage

```sql
CREATE STAGE stage_externo;
```

## File Formats

Permiten definir cómo debe interpretarse un archivo.

Ejemplo:

```sql
CREATE FILE FORMAT formato_csv
TYPE = CSV
FIELD_DELIMITER = ','
SKIP_HEADER = 1;
```

Beneficios:

* Reutilización.
* Consistencia.
* Simplificación de cargas.

## Virtual Warehouses

Los Virtual Warehouses son recursos de procesamiento independientes.

A diferencia de BigQuery, Snowflake requiere un Warehouse para ejecutar consultas.

Ejemplo:

```sql
CREATE WAREHOUSE etl_wh
WITH
WAREHOUSE_SIZE = 'SMALL';
```

Consulta utilizando el Warehouse:

```sql
USE WAREHOUSE etl_wh;

SELECT *
FROM clientes;
```

## Organización Empresarial Recomendada

Ejemplo:

```text
ventas_db
│
├── comercial
│   ├── clientes
│   ├── productos
│   └── ventas
│
├── reportes
│   ├── v_resumen_mensual
│   └── mv_top_productos
│
└── staging
    ├── stage_clientes
    └── stage_ventas
```

Warehouses:

```text
ETL_WH
BI_WH
DATA_SCIENCE_WH
```

## Diferencias con BigQuery

| Concepto               | BigQuery | Snowflake          |
| ---------------------- | -------- | ------------------ |
| Contenedor principal   | Project  | Database           |
| Organización lógica    | Dataset  | Schema             |
| Procesamiento          | Slots    | Virtual Warehouses |
| Área de carga          | Externa  | Stages             |
| Formatos reutilizables | Limitado | File Formats       |

## Buenas Prácticas

* Separar datos operativos y reportes en distintos Schemas.
* Utilizar Stages para organizar archivos de carga.
* Crear Warehouses específicos para cada tipo de proceso.
* Utilizar nombres consistentes para Databases y Schemas.
* Mantener los objetos organizados por dominio de negocio.

## Resumen

Snowflake organiza la información mediante Databases, Schemas, Tables y Views, incorporando además componentes propios como Stages, File Formats y Virtual Warehouses. Esta separación entre almacenamiento y procesamiento proporciona una gran flexibilidad para administrar cargas analíticas, ETL y ambientes empresariales de alta concurrencia.


# Carga de Datos en BigQuery

## Introducción

La carga de datos es una de las tareas más comunes dentro de BigQuery. En proyectos reales, los datos suelen provenir de archivos CSV, JSON, Parquet, sistemas transaccionales o servicios de almacenamiento en la nube.

BigQuery permite cargar información mediante la interfaz gráfica, comandos SQL, herramientas de línea de comandos y procesos automatizados.

## Métodos de Carga

Las formas más utilizadas para cargar datos son:

* Archivos locales.
* Google Cloud Storage.
* Consultas SQL.
* APIs.
* Herramientas ETL.
* Airflow.
* dbt.

## Formatos Soportados

BigQuery soporta múltiples formatos de archivo.

| Formato | Uso recomendado            |
| ------- | -------------------------- |
| CSV     | Datos tabulares simples    |
| JSON    | Datos semiestructurados    |
| Parquet | Grandes volúmenes de datos |
| Avro    | Intercambio de datos       |
| ORC     | Procesamiento distribuido  |

En ambientes productivos, Parquet suele ser la opción preferida por su eficiencia en almacenamiento y consultas.

## Carga desde un Archivo CSV

Supongamos el siguiente archivo:

**clientes.csv**

| id_cliente | nombre      | pais       |
| ---------- | ----------- | ---------- |
| 1          | Juan Pérez  | Guatemala  |
| 2          | Ana López   | México     |
| 3          | Carlos Ruiz | Costa Rica |

### Creación de la tabla

```sql
CREATE TABLE ventas.clientes (
    id_cliente INT64,
    nombre STRING,
    pais STRING
);
```

Posteriormente se selecciona el archivo desde la interfaz de BigQuery y se carga sobre la tabla.

## Carga desde Google Cloud Storage

En proyectos empresariales es común almacenar archivos en un bucket antes de cargarlos.

Ejemplo:

```text
gs://empresa-datalake/clientes/clientes.csv
```

Carga mediante SQL:

```sql
LOAD DATA INTO ventas.clientes
FROM FILES (
    format = 'CSV',
    uris = ['gs://empresa-datalake/clientes/clientes.csv']
);
```

## Carga de Archivos JSON

Archivo:

```json
[
  {
    "id_cliente":1,
    "nombre":"Juan Pérez",
    "pais":"Guatemala"
  },
  {
    "id_cliente":2,
    "nombre":"Ana López",
    "pais":"México"
  }
]
```

Tabla:

```sql
CREATE TABLE ventas.clientes_json (
    id_cliente INT64,
    nombre STRING,
    pais STRING
);
```

BigQuery detectará automáticamente la estructura del archivo.

## Carga de Archivos Parquet

Parquet es uno de los formatos más utilizados en Ingeniería de Datos.

Ventajas:

* Menor espacio de almacenamiento.
* Lectura columnar.
* Mayor velocidad de consulta.
* Integración con Spark y Data Lakes.

Ejemplo:

```text
gs://empresa-datalake/ventas/ventas.parquet
```

Carga:

```sql
LOAD DATA INTO ventas.ventas
FROM FILES (
    format = 'PARQUET',
    uris = ['gs://empresa-datalake/ventas/ventas.parquet']
);
```

## Autodetección de Esquema

BigQuery puede identificar automáticamente:

* Tipos de datos.
* Columnas.
* Longitudes.
* Fechas.

Ejemplo:

| Campo       | Tipo Detectado |
| ----------- | -------------- |
| id_cliente  | INT64          |
| nombre      | STRING         |
| fecha_venta | DATE           |
| monto       | NUMERIC        |

Aunque resulta conveniente para pruebas rápidas, en ambientes productivos se recomienda definir el esquema manualmente.

## Carga Mediante Consulta

También es posible crear tablas a partir de consultas.

Ejemplo:

```sql
CREATE TABLE ventas.resumen_ventas AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas.transacciones
GROUP BY producto;
```

Resultado:

| producto | total_ventas |
| -------- | ------------ |
| Laptop   | 50000        |
| Mouse    | 12000        |
| Monitor  | 30000        |

## Carga Incremental

En lugar de reemplazar toda la información, es común agregar únicamente nuevos registros.

Ejemplo:

```sql
INSERT INTO ventas.transacciones
VALUES
(101,'Laptop',4500),
(102,'Mouse',150);
```

Esto evita reprocesar grandes volúmenes de datos.

## Carga Mediante Airflow

Proceso típico:

```text
Archivo CSV
      ↓
Cloud Storage
      ↓
Airflow DAG
      ↓
BigQuery
      ↓
Dashboard
```

Este enfoque es ampliamente utilizado en arquitecturas modernas de datos.

## Caso Empresarial

Una empresa de retail recibe diariamente:

* Ventas.
* Inventarios.
* Productos.
* Clientes.

Cada noche se generan archivos Parquet:

```text
ventas_2026_06_15.parquet
clientes_2026_06_15.parquet
inventario_2026_06_15.parquet
```

Proceso:

```text
ERP
 ↓
Cloud Storage
 ↓
BigQuery
 ↓
Power BI / Looker
```

Los analistas pueden consultar la información al día siguiente sin intervención manual.

## Buenas Prácticas

* Utilizar Parquet para grandes volúmenes de datos.
* Definir esquemas explícitamente.
* Validar la calidad de los datos antes de cargarlos.
* Automatizar cargas mediante Airflow.
* Evitar cargas manuales en producción.
* Mantener una zona de almacenamiento intermedia en Cloud Storage.

## Errores Comunes

### Utilizar CSV para grandes volúmenes

Los archivos CSV ocupan más espacio y generan consultas menos eficientes.

### Confiar siempre en la autodetección

Puede interpretar incorrectamente fechas o números.

### Reemplazar tablas completas innecesariamente

Incrementa costos y tiempos de procesamiento.

### No validar datos antes de cargarlos

Provoca problemas posteriores en reportes y modelos analíticos.

## Resumen

BigQuery permite cargar datos desde múltiples fuentes y formatos, siendo Cloud Storage y Parquet la combinación más utilizada en ambientes empresariales. Aunque la plataforma facilita la carga mediante autodetección y herramientas gráficas, en producción se recomienda automatizar los procesos, definir esquemas explícitos y utilizar formatos optimizados para maximizar el rendimiento y reducir costos.


# Carga de Datos en Snowflake

## Introducción

La carga de datos en Snowflake gira alrededor de tres conceptos fundamentales:

* Stages
* File Formats
* COPY INTO

Estos componentes permiten construir procesos de ingestión altamente escalables y fáciles de automatizar.

A diferencia de BigQuery, donde normalmente se carga directamente desde Cloud Storage, Snowflake utiliza una capa intermedia denominada Stage para gestionar los archivos que serán procesados.

## Flujo General de Carga

Proceso típico:

```text
Archivo
   ↓
Stage
   ↓
COPY INTO
   ↓
Tabla Snowflake
```

Ejemplo:

```text
ventas.csv
    ↓
stage_ventas
    ↓
COPY INTO ventas
    ↓
Tabla final
```

## Archivos Soportados

Snowflake soporta múltiples formatos.

| Formato | Soporte |
| ------- | ------- |
| CSV     | Sí      |
| JSON    | Sí      |
| Parquet | Sí      |
| Avro    | Sí      |
| ORC     | Sí      |
| XML     | Sí      |

En ambientes productivos, Parquet suele ser la mejor opción por rendimiento y compresión.

## Creación de una Tabla

Antes de cargar información, se crea la tabla destino.

```sql
CREATE TABLE clientes (
    id_cliente INTEGER,
    nombre STRING,
    pais STRING
);
```

Resultado:

| id_cliente | nombre     | pais      |
| ---------- | ---------- | --------- |
| 1          | Juan Pérez | Guatemala |
| 2          | Ana López  | México    |

## Stages

Los Stages son ubicaciones temporales donde Snowflake accede a los archivos antes de cargarlos.

### Internal Stage

Los archivos son almacenados dentro de Snowflake.

```sql
CREATE STAGE stage_clientes;
```

Resultado:

```text
stage_clientes
```

Posteriormente se cargan archivos al Stage.

```sql
PUT file://clientes.csv @stage_clientes;
```

## Verificar Archivos del Stage

```sql
LIST @stage_clientes;
```

Resultado:

| Archivo      |
| ------------ |
| clientes.csv |

## External Stage

Permite trabajar con archivos almacenados fuera de Snowflake.

Ejemplos:

* Amazon S3
* Azure Blob Storage
* Google Cloud Storage

Ejemplo conceptual:

```sql
CREATE STAGE stage_ventas
URL='s3://empresa-data/ventas/';
```

Los archivos permanecen en el almacenamiento externo.

## File Formats

Definen cómo Snowflake debe interpretar los archivos.

### Formato CSV

```sql
CREATE FILE FORMAT formato_csv
TYPE = CSV
FIELD_DELIMITER = ','
SKIP_HEADER = 1;
```

### Formato JSON

```sql
CREATE FILE FORMAT formato_json
TYPE = JSON;
```

### Formato Parquet

```sql
CREATE FILE FORMAT formato_parquet
TYPE = PARQUET;
```

Una vez creado, el formato puede reutilizarse en múltiples cargas.

## COPY INTO

Es el comando principal para cargar datos.

Ejemplo:

```sql
COPY INTO clientes
FROM @stage_clientes
FILE_FORMAT = formato_csv;
```

Proceso:

```text
clientes.csv
      ↓
stage_clientes
      ↓
COPY INTO
      ↓
clientes
```

## Verificación de Datos

```sql
SELECT *
FROM clientes;
```

Resultado:

| id_cliente | nombre     | pais      |
| ---------- | ---------- | --------- |
| 1          | Juan Pérez | Guatemala |
| 2          | Ana López  | México    |

## Carga de Archivos Parquet

Archivo:

```text
ventas.parquet
```

Comando:

```sql
COPY INTO ventas
FROM @stage_ventas
FILE_FORMAT = formato_parquet;
```

Snowflake detecta automáticamente la estructura interna del archivo.

## Carga de Archivos JSON

Archivo:

```json
{
  "id_cliente":1,
  "nombre":"Juan Pérez",
  "pais":"Guatemala"
}
```

Comando:

```sql
COPY INTO clientes_json
FROM @stage_json
FILE_FORMAT = formato_json;
```

Este enfoque es ampliamente utilizado para APIs y datos semiestructurados.

## Transformaciones Durante la Carga

Snowflake permite transformar datos mientras se cargan.

Ejemplo:

```sql
COPY INTO clientes
FROM (
    SELECT
        $1,
        UPPER($2),
        $3
    FROM @stage_clientes
)
FILE_FORMAT = formato_csv;
```

Resultado:

| nombre original | nombre cargado |
| --------------- | -------------- |
| Juan Pérez      | JUAN PÉREZ     |
| Ana López       | ANA LÓPEZ      |

## Validación de Errores

Snowflake puede validar archivos antes de cargarlos.

```sql
COPY INTO clientes
FROM @stage_clientes
FILE_FORMAT = formato_csv
VALIDATION_MODE='RETURN_ERRORS';
```

Beneficios:

* Detectar errores de formato.
* Validar tipos de datos.
* Reducir fallos en producción.

## Carga Incremental

Escenario:

Cada día llega un nuevo archivo.

```text
ventas_2026_06_15.parquet
ventas_2026_06_16.parquet
ventas_2026_06_17.parquet
```

Snowflake puede cargar únicamente los archivos nuevos.

Esto evita reprocesamientos innecesarios.

## Snowpipe

Snowpipe permite automatizar la carga de datos casi en tiempo real.

Flujo:

```text
Archivo nuevo
      ↓
Cloud Storage
      ↓
Snowpipe
      ↓
Snowflake
```

Ventajas:

* Automatización.
* Menor intervención manual.
* Datos disponibles rápidamente.

## Caso Empresarial

Una empresa financiera recibe diariamente:

* Transacciones.
* Clientes.
* Movimientos bancarios.

Proceso:

```text
Sistema Core
      ↓
Parquet
      ↓
AWS S3
      ↓
External Stage
      ↓
COPY INTO
      ↓
Snowflake
      ↓
Power BI
```

Los reportes se actualizan automáticamente cada mañana.

## Comparación con BigQuery

| Característica               | BigQuery       | Snowflake   |
| ---------------------------- | -------------- | ----------- |
| Área temporal de carga       | No obligatoria | Stage       |
| Configuración de formato     | Opcional       | File Format |
| Comando principal            | LOAD DATA      | COPY INTO   |
| Carga automática             | Data Transfer  | Snowpipe    |
| Transformación durante carga | Limitada       | Avanzada    |

## Buenas Prácticas

* Utilizar Parquet para grandes volúmenes.
* Separar Stages por dominio de negocio.
* Reutilizar File Formats.
* Validar archivos antes de cargar.
* Automatizar procesos mediante Snowpipe.
* Mantener control de versiones sobre los archivos fuente.

## Resumen

La carga de datos en Snowflake se basa principalmente en el uso de Stages, File Formats y COPY INTO. Esta arquitectura proporciona una gran flexibilidad para trabajar con archivos provenientes de múltiples fuentes y formatos. Además, herramientas como Snowpipe permiten automatizar procesos de ingestión, convirtiendo a Snowflake en una plataforma altamente eficiente para entornos empresariales modernos.


# Particionamiento y Clustering en BigQuery



## Introducción

Uno de los errores más costosos en BigQuery es consultar tablas grandes sin utilizar particionamiento o clustering. Debido a que BigQuery cobra principalmente por la cantidad de datos procesados, una mala estrategia de diseño puede incrementar significativamente los costos y afectar el rendimiento.

Por esta razón, el particionamiento y el clustering son dos de las técnicas más importantes para optimizar consultas en BigQuery.

## ¿Qué es el Particionamiento?

El particionamiento consiste en dividir una tabla en segmentos más pequeños utilizando una columna específica.

En lugar de recorrer toda la tabla, BigQuery lee únicamente las particiones necesarias.

### Sin Particionamiento

Tabla:

```text
ventas
```

Registros:

| Año  |
| ---- |
| 2022 |
| 2023 |
| 2024 |
| 2025 |
| 2026 |

Consulta:

```sql
SELECT *
FROM ventas
WHERE fecha_venta = '2026-06-15';
```

BigQuery podría analizar toda la tabla.

## Con Particionamiento

Tabla:

```text
ventas_particionadas
```

Particiones:

```text
2022
2023
2024
2025
2026
```

Consulta:

```sql
SELECT *
FROM ventas_particionadas
WHERE fecha_venta = '2026-06-15';
```

BigQuery leerá únicamente la partición correspondiente.

## Beneficios del Particionamiento

* Menor cantidad de datos procesados.
* Menor costo de consulta.
* Mejor rendimiento.
* Mayor escalabilidad.

## Tipos de Particionamiento

### Particionamiento por Fecha

Es el más utilizado.

Ejemplo:

```sql
CREATE TABLE ventas (
    id_venta INT64,
    fecha_venta DATE,
    monto NUMERIC
)
PARTITION BY fecha_venta;
```

Resultado conceptual:

```text
ventas
│
├── 2026-06-13
├── 2026-06-14
├── 2026-06-15
└── 2026-06-16
```

## Particionamiento por Timestamp

```sql
CREATE TABLE eventos (
    id_evento INT64,
    fecha_evento TIMESTAMP
)
PARTITION BY DATE(fecha_evento);
```

## Particionamiento por Enteros

Útil para ciertos escenarios específicos.

```sql
CREATE TABLE clientes (
    id_cliente INT64,
    nombre STRING
)
PARTITION BY RANGE_BUCKET(
    id_cliente,
    GENERATE_ARRAY(0,1000000,100000)
);
```

## Buenas Prácticas para Particiones

Utilizar columnas que:

* Se consulten frecuentemente.
* Tengan distribución uniforme.
* Permitan filtrar información.

Ejemplos comunes:

| Columna           | Recomendado |
| ----------------- | ----------- |
| fecha_venta       | Sí          |
| fecha_transaccion | Sí          |
| fecha_registro    | Sí          |
| nombre_cliente    | No          |
| correo            | No          |

## ¿Qué es el Clustering?

El Clustering organiza físicamente los datos dentro de una partición utilizando una o varias columnas.

Mientras el particionamiento reduce la cantidad de particiones leídas, el clustering reduce la cantidad de bloques leídos dentro de cada partición.

## Ejemplo de Clustering

Tabla:

```sql
CREATE TABLE ventas
(
    id_venta INT64,
    fecha_venta DATE,
    pais STRING,
    producto STRING,
    monto NUMERIC
)
CLUSTER BY pais;
```

BigQuery organiza internamente los registros por país.

Resultado conceptual:

```text
Guatemala
Guatemala
Guatemala
México
México
México
Costa Rica
Costa Rica
```

## Consulta Optimizada

```sql
SELECT *
FROM ventas
WHERE pais = 'Guatemala';
```

BigQuery encontrará los registros mucho más rápido.

## Clustering con Varias Columnas

```sql
CREATE TABLE ventas
(
    id_venta INT64,
    fecha_venta DATE,
    pais STRING,
    producto STRING,
    monto NUMERIC
)
CLUSTER BY pais, producto;
```

Orden conceptual:

```text
Guatemala → Laptop
Guatemala → Mouse
Guatemala → Monitor

México → Laptop
México → Mouse
México → Monitor
```

## Particionamiento y Clustering Combinados

En ambientes productivos es común utilizar ambos.

Ejemplo:

```sql
CREATE TABLE ventas
(
    id_venta INT64,
    fecha_venta DATE,
    pais STRING,
    producto STRING,
    monto NUMERIC
)
PARTITION BY fecha_venta
CLUSTER BY pais;
```

Estructura conceptual:

```text
2026-06-14
    ├── Guatemala
    ├── México
    └── Costa Rica

2026-06-15
    ├── Guatemala
    ├── México
    └── Costa Rica
```

## Comparación de Rendimiento

Tabla:

```text
ventas
```

Registros:

```text
500 millones
```

Consulta:

```sql
SELECT *
FROM ventas
WHERE fecha_venta = '2026-06-15'
AND pais = 'Guatemala';
```

| Diseño                 | Datos Procesados |
| ---------------------- | ---------------- |
| Sin optimización       | 500 millones     |
| Particionado           | 15 millones      |
| Particionado + Cluster | 1 millón         |

*Valores ilustrativos para comprender el impacto.*

## Requerir Filtros de Partición

BigQuery permite obligar a los usuarios a utilizar filtros de partición.

```sql
ALTER TABLE ventas
SET OPTIONS (
    require_partition_filter = TRUE
);
```

Consulta correcta:

```sql
SELECT *
FROM ventas
WHERE fecha_venta = '2026-06-15';
```

Consulta incorrecta:

```sql
SELECT *
FROM ventas;
```

Resultado:

```text
Error
Partition filter required
```

Esta práctica ayuda a controlar costos.

## Caso Empresarial

Una cadena de supermercados almacena:

```text
5 mil millones de registros
```

Tabla:

```text
ventas
```

Diseño recomendado:

```sql
PARTITION BY fecha_venta
CLUSTER BY sucursal, producto
```

Consultas frecuentes:

```sql
SELECT *
FROM ventas
WHERE fecha_venta BETWEEN '2026-06-01'
AND '2026-06-30'
AND sucursal = 'Guatemala Norte';
```

Beneficios:

* Menor costo.
* Consultas más rápidas.
* Menor consumo de recursos.

## Errores Comunes

### No particionar tablas históricas

Provoca consultas costosas y lentas.

### Particionar por columnas incorrectas

Ejemplo:

```sql
PARTITION BY nombre_cliente
```

No aporta beneficios reales.

### Crear demasiadas columnas de clustering

El exceso de columnas puede reducir la efectividad del clustering.

### No utilizar filtros en las consultas

Aunque exista particionamiento, si la consulta no filtra adecuadamente se perderán sus beneficios.

## Recomendaciones para Producción

| Situación                         | Recomendación        |
| --------------------------------- | -------------------- |
| Datos históricos                  | Particionar          |
| Datos transaccionales             | Particionar          |
| Consultas frecuentes por país     | Cluster              |
| Consultas frecuentes por producto | Cluster              |
| Tablas pequeñas                   | No siempre necesario |

## Resumen

El particionamiento y el clustering son dos de las herramientas más importantes para optimizar BigQuery. El particionamiento reduce la cantidad de datos analizados al dividir las tablas en segmentos más pequeños, mientras que el clustering organiza los datos dentro de cada partición para acelerar las búsquedas. Utilizados correctamente, pueden reducir drásticamente los costos de consulta y mejorar significativamente el rendimiento en ambientes productivos.


# Optimización de Costos en BigQuery

## Introducción

Uno de los aspectos más importantes al trabajar con BigQuery es comprender cómo se generan los costos.

A diferencia de muchas bases de datos tradicionales, BigQuery cobra principalmente por la cantidad de datos procesados durante las consultas. Esto significa que dos consultas que devuelven el mismo resultado pueden tener costos muy diferentes dependiendo de cómo estén escritas.

Por esta razón, optimizar consultas no solo mejora el rendimiento, sino que también reduce significativamente los gastos operativos.

## ¿Cómo Cobra BigQuery?

Los costos principales provienen de:

* Almacenamiento de datos.
* Datos procesados por consultas.
* Streaming de datos (cuando aplica).
* Reservas de Slots (en algunos modelos empresariales).

Sin embargo, en la mayoría de los proyectos, el mayor impacto económico proviene de las consultas SQL.

## Evitar SELECT *

Uno de los errores más comunes es consultar todas las columnas cuando solo se necesitan algunas.

Consulta poco eficiente:

```sql
SELECT *
FROM ventas;
```

Consulta optimizada:

```sql
SELECT
    fecha_venta,
    producto,
    monto
FROM ventas;
```

### Beneficio

BigQuery leerá únicamente las columnas necesarias.

Menos datos procesados significa menor costo.

## Utilizar Particiones

Supongamos una tabla con cinco años de información.

Consulta poco eficiente:

```sql
SELECT *
FROM ventas
WHERE producto = 'Laptop';
```

Consulta optimizada:

```sql
SELECT *
FROM ventas
WHERE fecha_venta BETWEEN '2026-06-01'
AND '2026-06-30'
AND producto = 'Laptop';
```

Si la tabla está particionada por fecha, BigQuery leerá únicamente el período solicitado.

## Habilitar Partition Filter

Es recomendable obligar a los usuarios a filtrar por partición.

```sql
ALTER TABLE ventas
SET OPTIONS (
    require_partition_filter = TRUE
);
```

Beneficio:

* Previene consultas costosas.
* Reduce errores de los usuarios.
* Facilita el control presupuestario.

## Utilizar Clustering

Cuando las consultas filtran frecuentemente por ciertas columnas, el clustering puede reducir considerablemente la cantidad de datos analizados.

Ejemplo:

```sql
CREATE TABLE ventas
PARTITION BY fecha_venta
CLUSTER BY pais;
```

Consulta frecuente:

```sql
SELECT *
FROM ventas
WHERE pais = 'Guatemala';
```

BigQuery localizará la información de forma más eficiente.

## Evitar Consultas Repetitivas

Situación común:

```sql
SELECT
    producto,
    SUM(monto)
FROM ventas
GROUP BY producto;
```

Ejecutada cientos de veces al día.

Solución:

Materialized View.

```sql
CREATE MATERIALIZED VIEW mv_ventas_producto AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas
GROUP BY producto;
```

Beneficios:

* Menor procesamiento.
* Menor tiempo de respuesta.
* Menor costo.

## Utilizar Tablas de Resumen

En lugar de consultar tablas gigantes constantemente, es recomendable crear agregaciones periódicas.

Tabla original:

```text
ventas_detalle
```

5 mil millones de registros.

Tabla resumen:

```text
ventas_mensuales
```

50 mil registros.

Consulta:

```sql
SELECT *
FROM ventas_mensuales;
```

Mucho más económica que consultar el detalle completo.

## Revisar el Volumen Procesado

Antes de ejecutar una consulta, BigQuery muestra una estimación de datos procesados.

Ejemplo:

| Consulta   | Datos Procesados |
| ---------- | ---------------- |
| Consulta A | 20 GB            |
| Consulta B | 2 TB             |

Esta información debe revisarse antes de ejecutar consultas complejas.

## Utilizar Tablas Expirables

Para datos temporales:

```sql
CREATE TABLE tabla_temporal
(
    id INT64
)
OPTIONS (
    expiration_timestamp =
    TIMESTAMP_ADD(CURRENT_TIMESTAMP(), INTERVAL 7 DAY)
);
```

Beneficios:

* Menor almacenamiento.
* Limpieza automática.
* Menor administración.

## Evitar Duplicación de Datos

Mala práctica:

```text
ventas_2024
ventas_2025
ventas_2026
ventas_backup
ventas_backup2
```

Buena práctica:

```text
ventas
```

Particionada por fecha.

Beneficios:

* Menor almacenamiento.
* Menor complejidad.
* Mejor mantenimiento.

## Optimizar JOINs

Consulta poco eficiente:

```sql
SELECT *
FROM ventas v
JOIN clientes c
ON v.id_cliente = c.id_cliente;
```

Consulta optimizada:

```sql
SELECT
    v.id_venta,
    c.nombre
FROM ventas v
JOIN clientes c
ON v.id_cliente = c.id_cliente;
```

Seleccionar únicamente las columnas necesarias reduce el volumen procesado.

## Utilizar Aproximaciones Cuando Sea Posible

Consulta exacta:

```sql
SELECT COUNT(DISTINCT id_cliente)
FROM ventas;
```

Consulta aproximada:

```sql
SELECT APPROX_COUNT_DISTINCT(id_cliente)
FROM ventas;
```

Ventajas:

* Menor tiempo de ejecución.
* Menor consumo de recursos.
* Resultados adecuados para análisis exploratorios.

## Monitoreo de Costos

BigQuery permite consultar información histórica.

Ejemplo:

```sql
SELECT *
FROM region-us.INFORMATION_SCHEMA.JOBS;
```

Información disponible:

* Usuario.
* Tiempo de ejecución.
* Datos procesados.
* Costos estimados.

Esto permite identificar consultas problemáticas.

## Caso Empresarial

Una empresa ejecuta diariamente:

```text
800 reportes
```

Consulta original:

```sql
SELECT *
FROM ventas_historicas;
```

Datos procesados:

```text
5 TB diarios
```

Después de aplicar:

* Particiones.
* Clustering.
* Materialized Views.

Nuevo consumo:

```text
400 GB diarios
```

Resultado:

| Métrica            | Antes    | Después   |
| ------------------ | -------- | --------- |
| Datos procesados   | 5 TB     | 400 GB    |
| Tiempo de consulta | Alto     | Bajo      |
| Costos             | Elevados | Reducidos |

## Checklist para Producción

Antes de ejecutar una consulta verificar:

* ¿Se están seleccionando solo las columnas necesarias?
* ¿Existe filtro de partición?
* ¿Se puede usar clustering?
* ¿Existe una Materialized View disponible?
* ¿Se está consultando una tabla resumen?
* ¿Se revisó la estimación de datos procesados?

## Buenas Prácticas

* Evitar SELECT *.
* Utilizar particionamiento siempre que sea posible.
* Implementar clustering en tablas grandes.
* Crear Materialized Views para consultas recurrentes.
* Monitorear consultas costosas.
* Revisar periódicamente los patrones de consumo.

## Resumen

La optimización de costos en BigQuery está directamente relacionada con la cantidad de datos procesados. Técnicas como el particionamiento, clustering, uso de Materialized Views, tablas resumen y selección adecuada de columnas permiten reducir significativamente los costos y mejorar el rendimiento. Un Data Engineer debe considerar estos aspectos desde el diseño inicial de la solución para evitar gastos innecesarios en ambientes productivos.


# Optimización de Costos en Snowflake

## Introducción

Uno de los errores más comunes al comenzar a trabajar con Snowflake es asumir que los costos funcionan igual que en BigQuery.

Mientras que BigQuery se enfoca principalmente en los datos procesados, Snowflake basa gran parte de su modelo de costos en el consumo de cómputo a través de los Virtual Warehouses.

Por esta razón, una mala administración de los Warehouses puede generar costos innecesarios incluso cuando no se están ejecutando consultas.

## Componentes Principales del Costo

Los costos en Snowflake se dividen principalmente en:

* Almacenamiento.
* Cómputo (Virtual Warehouses).
* Servicios Cloud (generalmente de menor impacto).

En la mayoría de los proyectos, el mayor gasto proviene del cómputo.

## Comprendiendo los Créditos

Snowflake utiliza una unidad llamada crédito para medir el consumo de recursos.

Cada Warehouse consume créditos según su tamaño y tiempo de ejecución.

Ejemplo conceptual:

| Warehouse | Consumo Relativo |
| --------- | ---------------- |
| X-Small   | Bajo             |
| Small     | Medio            |
| Medium    | Alto             |
| Large     | Muy Alto         |
| X-Large   | Elevado          |

A medida que aumenta el tamaño del Warehouse, también aumenta el costo operativo.

## Seleccionar el Tamaño Correcto

Error común:

Utilizar un Warehouse Large para consultas simples.

```sql
SELECT *
FROM clientes;
```

No existe beneficio real en este escenario.

Buena práctica:

| Tipo de Trabajo  | Tamaño Recomendado |
| ---------------- | ------------------ |
| Desarrollo       | X-Small            |
| Reportería       | Small              |
| ETL moderado     | Medium             |
| Procesos masivos | Large              |

Siempre se recomienda comenzar con tamaños pequeños y escalar únicamente cuando sea necesario.

## Auto Suspend

Una de las configuraciones más importantes en Snowflake.

Permite apagar automáticamente un Warehouse después de cierto tiempo de inactividad.

Ejemplo:

```sql
ALTER WAREHOUSE BI_WH
SET AUTO_SUSPEND = 300;
```

Resultado:

```text
5 minutos sin actividad
↓
Warehouse suspendido
↓
No consume créditos
```

## Auto Resume

Permite encender automáticamente un Warehouse cuando llega una nueva consulta.

```sql
ALTER WAREHOUSE BI_WH
SET AUTO_RESUME = TRUE;
```

Proceso:

```text
Usuario ejecuta consulta
↓
Warehouse apagado
↓
Snowflake lo enciende automáticamente
↓
Consulta ejecutada
```

Beneficio:

* Menor costo.
* Mejor experiencia de usuario.

## Evitar Warehouses Siempre Activos

Mala práctica:

```text
ETL_WH
24 horas encendido
```

Aunque no existan consultas, el Warehouse seguirá consumiendo recursos.

Buena práctica:

```text
ETL_WH
Auto Suspend = 300 segundos
```

El Warehouse solo consume cuando realmente se utiliza.

## Separar Cargas de Trabajo

Escenario incorrecto:

```text
WH_GENERAL
```

Utilizado por:

* BI
* ETL
* Data Science
* Analistas

Problemas:

* Competencia por recursos.
* Consultas más lentas.
* Escalamiento innecesario.

Escenario recomendado:

```text
BI_WH
ETL_WH
DS_WH
```

Cada equipo utiliza recursos independientes.

## Multi-Cluster Warehouses

Cuando existe alta concurrencia, Snowflake puede crear múltiples clústeres automáticamente.

Ejemplo:

```text
100 usuarios simultáneos
```

Configuración:

```text
Min Cluster = 1
Max Cluster = 3
```

Beneficio:

* Mejor rendimiento.
* Escalamiento automático.

Consideración:

* Mayor costo si no se administra correctamente.

## Monitorear Consultas Costosas

Snowflake proporciona historial de consultas.

Ejemplo:

```sql
SELECT *
FROM SNOWFLAKE.ACCOUNT_USAGE.QUERY_HISTORY;
```

Información disponible:

* Usuario.
* Tiempo de ejecución.
* Warehouse utilizado.
* Recursos consumidos.

Esto permite identificar consultas ineficientes.

## Evitar Procesamientos Innecesarios

Consulta poco eficiente:

```sql
SELECT *
FROM ventas;
```

Consulta optimizada:

```sql
SELECT
    fecha_venta,
    producto,
    monto
FROM ventas;
```

Aunque Snowflake maneja eficientemente los datos, siempre es recomendable reducir el volumen procesado.

## Utilizar Materialized Views

Consulta frecuente:

```sql
SELECT
    producto,
    SUM(monto)
FROM ventas
GROUP BY producto;
```

Solución:

```sql
CREATE MATERIALIZED VIEW mv_ventas_producto AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas
GROUP BY producto;
```

Beneficios:

* Menor tiempo de respuesta.
* Menor consumo de cómputo.
* Menor carga sobre los Warehouses.

## Suspender Warehouses de Desarrollo

En muchos proyectos los mayores desperdicios ocurren en ambientes de desarrollo.

Ejemplo:

```text
DEV_WH
```

Encendido:

```text
24 horas al día
```

Uso real:

```text
2 horas al día
```

Implementando Auto Suspend:

* Menor consumo.
* Menor facturación.

## Monitoreo de Consumo

Snowflake proporciona métricas sobre utilización de Warehouses.

Información relevante:

* Tiempo activo.
* Créditos consumidos.
* Usuarios conectados.
* Consultas ejecutadas.

Esto permite detectar recursos sobredimensionados.

## Caso Empresarial

Una empresa posee:

| Warehouse | Tamaño |
| --------- | ------ |
| BI_WH     | Medium |
| ETL_WH    | Large  |
| DS_WH     | Large  |

Situación inicial:

```text
24 horas activos
```

Problema:

* Alto consumo de créditos.

Optimización aplicada:

```text
Auto Suspend
Auto Resume
Separación de cargas
Monitoreo periódico
```

Resultado:

| Métrica             | Antes    | Después    |
| ------------------- | -------- | ---------- |
| Consumo de créditos | Alto     | Reducido   |
| Rendimiento         | Variable | Estable    |
| Administración      | Compleja | Controlada |

## Checklist para Producción

Antes de crear un Warehouse verificar:

* ¿Realmente necesita ese tamaño?
* ¿Tiene Auto Suspend habilitado?
* ¿Tiene Auto Resume habilitado?
* ¿Existe monitoreo de consumo?
* ¿Está separado por tipo de carga?
* ¿Se revisan consultas costosas regularmente?

## Buenas Prácticas

* Comenzar con Warehouses pequeños.
* Escalar únicamente cuando sea necesario.
* Configurar Auto Suspend.
* Configurar Auto Resume.
* Separar cargas de trabajo.
* Monitorear consumo de créditos.
* Revisar periódicamente el historial de consultas.

## Resumen

La optimización de costos en Snowflake se centra principalmente en la administración eficiente de los Virtual Warehouses. Configuraciones como Auto Suspend y Auto Resume, junto con una correcta selección del tamaño de los Warehouses y la separación de cargas de trabajo, permiten reducir significativamente el consumo de créditos. Un monitoreo constante y una estrategia adecuada de cómputo son fundamentales para mantener una plataforma escalable y económicamente eficiente.


# Seguridad en BigQuery

## Introducción

La seguridad en BigQuery se basa principalmente en el sistema de Identity and Access Management (IAM) de Google Cloud. Mediante IAM es posible controlar quién puede acceder a los recursos, qué acciones puede realizar y sobre qué objetos tiene permisos.

En ambientes empresariales, una correcta configuración de permisos es fundamental para proteger información sensible y cumplir con políticas de gobernanza de datos.

## Niveles de Seguridad en BigQuery

Los permisos pueden asignarse en distintos niveles:

```text
Organización
    ↓
Proyecto
    ↓
Dataset
    ↓
Tabla
    ↓
Vista
```

Mientras más específico sea el nivel, mayor control se obtiene sobre los accesos.

## Conceptos Fundamentales

### Principal

Es la identidad que recibe permisos.

Ejemplos:

* Usuarios.
* Grupos.
* Service Accounts.
* Aplicaciones.

### Rol

Es un conjunto de permisos predefinidos.

Ejemplo:

```text
BigQuery Data Viewer
```

Permite consultar datos, pero no modificarlos.

### Permiso

Es una acción específica.

Ejemplos:

```text
bigquery.tables.get
bigquery.tables.create
bigquery.jobs.create
```

## Roles Más Utilizados

### BigQuery Data Viewer

Permite:

* Consultar tablas.
* Consultar vistas.
* Leer información.

No permite:

* Crear tablas.
* Modificar datos.
* Eliminar objetos.

### BigQuery Data Editor

Permite:

* Insertar datos.
* Actualizar registros.
* Crear tablas.

No permite administrar permisos.

### BigQuery Data Owner

Permite:

* Administrar tablas.
* Modificar permisos.
* Eliminar objetos.

Debe utilizarse con precaución.

### BigQuery Job User

Permite ejecutar consultas.

Es uno de los roles más utilizados para analistas.

## Asignación de Permisos

Ejemplo conceptual:

| Usuario       | Rol         |
| ------------- | ----------- |
| Analista BI   | Data Viewer |
| Data Engineer | Data Editor |
| Administrador | Data Owner  |

Aplicando el principio de mínimo privilegio.

## Seguridad a Nivel de Dataset

Es posible otorgar acceso únicamente a un Dataset específico.

Ejemplo:

```text
ventas
```

Acceso permitido:

```text
equipo_bi@empresa.com
```

Acceso denegado:

```text
equipo_marketing@empresa.com
```

Esto permite aislar dominios de negocio.

## Seguridad a Nivel de Tabla

También es posible restringir tablas específicas.

Ejemplo:

Dataset:

```text
rrhh
```

Tablas:

```text
empleados
nominas
evaluaciones
```

Solo el departamento de Recursos Humanos debería acceder a estas tablas.

## Seguridad Mediante Views

Una práctica muy utilizada consiste en ocultar información sensible mediante vistas.

Tabla original:

Vista para gerencia:

```sql
SELECT
    id_empleado,
    nombre
FROM empleados;
```

Los usuarios consultan la vista y nunca acceden directamente a la tabla.

Sin necesidad de duplicar tablas.

Permite restringir columnas sensibles.

Usuario estándar:

Mayor control sobre información confidencial.

* dbt.

* APIs.

* Consulta ejecutada.
* Recursos accedidos.

Esto permite:

```text
Todos son administradores
```

## Buenas Prácticas

* Aplicar el principio de mínimo privilegio.

* Implementar Row-Level Security cuando sea necesario.
* Utilizar Column-Level Security para datos sensibles.

## Resumen

La seguridad en BigQuery se basa en IAM y permite controlar el acceso a proyectos, datasets, tablas y vistas mediante roles y permisos. Funcionalidades como Row-Level Security, Column-Level Security y Service Accounts proporcionan mecanismos avanzados para proteger la información y cumplir con políticas de gobernanza de datos en ambientes empresariales.


# Seguridad en Snowflake

## Introducción

La seguridad en Snowflake se basa en un modelo denominado RBAC (Role-Based Access Control), considerado una de las características más sólidas de la plataforma.

A diferencia de otros sistemas donde los permisos suelen asignarse directamente a los usuarios, en Snowflake los permisos se asignan principalmente a roles, y posteriormente los roles se asignan a los usuarios.

Este enfoque facilita la administración de accesos en organizaciones con cientos o miles de usuarios.

## ¿Qué es RBAC?

RBAC significa:

```text
Role-Based Access Control
```

Control de Acceso Basado en Roles.

La idea principal es:

```text
Permisos
    ↓
Roles
    ↓
Usuarios
```

De esta forma los permisos se administran centralizadamente.

## Componentes Principales

### Usuarios

Representan personas o aplicaciones que acceden a Snowflake.

Ejemplos:

```text
juan.perez
ana.lopez
airflow_service
```

### Roles

Agrupan permisos relacionados.

Ejemplos:

```text
ANALYST_ROLE
DATA_ENGINEER_ROLE
ADMIN_ROLE
```

### Objetos

Son los recursos sobre los cuales se aplican permisos.

Ejemplos:

* Databases
* Schemas
* Tables
* Views
* Warehouses
* Stages

## Flujo de Permisos

Ejemplo:

```text
Tabla Ventas
       ↓
SELECT
       ↓
ANALYST_ROLE
       ↓
Juan Pérez
```

Juan podrá consultar la tabla porque hereda los permisos del rol.

## Creación de Roles

```sql
CREATE ROLE ANALYST_ROLE;
```

```sql
CREATE ROLE DATA_ENGINEER_ROLE;
```

```sql
CREATE ROLE ADMIN_ROLE;
```

## Asignación de Permisos

Permitir lectura sobre una tabla:

```sql
GRANT SELECT
ON TABLE ventas
TO ROLE ANALYST_ROLE;
```

Permitir inserciones:

```sql
GRANT INSERT
ON TABLE ventas
TO ROLE DATA_ENGINEER_ROLE;
```

## Asignación de Roles a Usuarios

```sql
GRANT ROLE ANALYST_ROLE
TO USER juan;
```

Resultado:

```text
Juan
 ↓
ANALYST_ROLE
 ↓
SELECT ventas
```

## Roles Más Utilizados

### SYSADMIN

Administra objetos de la plataforma.

Puede:

* Crear bases de datos.
* Crear tablas.
* Crear warehouses.

### SECURITYADMIN

Administra usuarios y permisos.

Puede:

* Crear usuarios.
* Crear roles.
* Asignar permisos.

### ACCOUNTADMIN

Es el rol con mayores privilegios.

Tiene acceso prácticamente completo a la cuenta.

Debe utilizarse únicamente para tareas administrativas específicas.

## Jerarquía de Roles

Snowflake permite que un rol herede permisos de otro.

Ejemplo:

```text
ANALYST_ROLE
        ↓
SENIOR_ANALYST_ROLE
        ↓
BI_MANAGER_ROLE
```

Los permisos se heredan automáticamente.

Beneficios:

* Menor administración.
* Mayor escalabilidad.
* Mejor organización.

## Permisos sobre Warehouses

Además de los datos, los usuarios necesitan acceso al cómputo.

Ejemplo:

```sql
GRANT USAGE
ON WAREHOUSE BI_WH
TO ROLE ANALYST_ROLE;
```

Sin este permiso el usuario no podrá ejecutar consultas.

## Permisos sobre Databases

```sql
GRANT USAGE
ON DATABASE ventas_db
TO ROLE ANALYST_ROLE;
```

Permite acceder a la base de datos.

## Permisos sobre Schemas

```sql
GRANT USAGE
ON SCHEMA ventas_db.comercial
TO ROLE ANALYST_ROLE;
```

Permite acceder al esquema.

## Permisos sobre Tablas

```sql
GRANT SELECT
ON TABLE clientes
TO ROLE ANALYST_ROLE;
```

Permite consultar la información.

## Flujo Completo de Acceso

```text
Warehouse
     ↓
Database
     ↓
Schema
     ↓
Table
     ↓
Role
     ↓
User
```

Si alguno de estos permisos falta, el acceso será denegado.

## Future Grants

Permiten aplicar permisos automáticamente a objetos futuros.

Ejemplo:

```sql
GRANT SELECT
ON FUTURE TABLES
IN SCHEMA ventas_db.comercial
TO ROLE ANALYST_ROLE;
```

Beneficio:

Cuando se cree una nueva tabla, el rol recibirá acceso automáticamente.

Evita configuraciones manuales repetitivas.

## Seguridad para Procesos Automatizados

Es recomendable crear usuarios específicos para herramientas como:

* Airflow
* dbt
* ETL
* APIs

Ejemplo:

```text
AIRFLOW_USER
```

Rol:

```text
AIRFLOW_ROLE
```

Beneficios:

* Mayor control.
* Auditoría más clara.
* Menor riesgo de seguridad.

## Auditoría

Snowflake registra actividades relacionadas con:

* Usuarios.
* Consultas.
* Roles.
* Permisos.
* Accesos.

Información disponible:

* Quién ejecutó una consulta.
* Cuándo la ejecutó.
* Qué datos consultó.
* Qué Warehouse utilizó.

Esto facilita el cumplimiento de políticas corporativas y normativas regulatorias.

## Caso Empresarial

Una empresa posee tres equipos:

### BI

Permisos:

```text
SELECT
```

Rol:

```text
BI_ANALYST_ROLE
```

### Ingeniería de Datos

Permisos:

```text
SELECT
INSERT
UPDATE
CREATE TABLE
```

Rol:

```text
DATA_ENGINEER_ROLE
```

### Administración

Permisos:

```text
Control total
```

Rol:

```text
ADMIN_ROLE
```

Resultado:

* Mejor gobernanza.
* Menor riesgo.
* Administración simplificada.

## Comparación con BigQuery

| Característica             | BigQuery  | Snowflake       |
| -------------------------- | --------- | --------------- |
| Modelo principal           | IAM       | RBAC            |
| Administración de permisos | Roles IAM | Roles Snowflake |
| Herencia de permisos       | Limitada  | Amplia          |
| Future Grants              | No        | Sí              |
| Flexibilidad               | Alta      | Muy alta        |

## Buenas Prácticas

* Asignar permisos a roles, no directamente a usuarios.
* Aplicar el principio de mínimo privilegio.
* Utilizar Future Grants cuando sea apropiado.
* Separar roles de negocio y roles técnicos.
* Utilizar cuentas específicas para automatizaciones.
* Auditar periódicamente usuarios y permisos.
* Limitar el uso de ACCOUNTADMIN.

## Resumen

Snowflake utiliza un modelo de seguridad basado en RBAC que permite administrar permisos de forma centralizada y escalable. Mediante roles, grants y herencia de permisos, es posible controlar el acceso a datos, warehouses y recursos de la plataforma de manera eficiente. Esta arquitectura facilita la gobernanza y convierte a Snowflake en una solución especialmente adecuada para organizaciones con múltiples equipos y grandes volúmenes de usuarios.


# Time Travel y Fail-Safe en Snowflake

## Introducción

Una de las características más diferenciadoras de Snowflake es su capacidad para recuperar datos históricos sin necesidad de restaurar respaldos completos.

Gracias a las funcionalidades de Time Travel y Fail-Safe, es posible recuperar información eliminada o modificada accidentalmente, reduciendo considerablemente el riesgo operativo en ambientes productivos.

## ¿Qué es Time Travel?

Time Travel permite consultar, restaurar o recuperar datos de un momento específico en el pasado.

Esto significa que, aunque una tabla haya sido modificada o eliminada, Snowflake conserva versiones históricas durante un período determinado.

## Casos de Uso Comunes

### Eliminación accidental

Supongamos que un usuario ejecuta:

```sql id="ywbqot"
DELETE FROM clientes;
```

Resultado:

```text id="clmz0h"
0 registros visibles
```

Sin Time Travel, la recuperación podría requerir respaldos externos.

Con Time Travel, los datos pueden recuperarse fácilmente.

### Actualización incorrecta

```sql id="b8q8qt"
UPDATE clientes
SET pais = 'Guatemala';
```

Resultado:

```text id="jgmf5n"
Todos los registros fueron modificados
```

Time Travel permite regresar a una versión anterior.

## Período de Retención

Dependiendo de la configuración y edición de Snowflake, los datos pueden mantenerse durante varios días.

Conceptualmente:

| Estado                         | Disponible |
| ------------------------------ | ---------- |
| Datos actuales                 | Sí         |
| Datos históricos               | Sí         |
| Datos eliminados recientemente | Sí         |

Durante este período se pueden recuperar versiones anteriores.

## Consultar Datos Históricos

Snowflake permite consultar una tabla en un momento específico.

Ejemplo conceptual:

```sql id="p8w7eg"
SELECT *
FROM ventas
AT (TIMESTAMP => '2026-06-15 10:00:00');
```

Resultado:

```text id="8krvnp"
Estado exacto de la tabla
a las 10:00 AM
```

Esto resulta extremadamente útil para auditorías e investigaciones.

## Recuperar una Tabla Eliminada

Supongamos:

```sql id="c5dqwl"
DROP TABLE clientes;
```

La tabla desaparece.

Sin embargo, puede recuperarse mediante:

```sql id="p32z6t"
UNDROP TABLE clientes;
```

Resultado:

```text id="z7ulv4"
Tabla restaurada
```

Esta es una de las funcionalidades más utilizadas en entornos empresariales.

## Consultar Versiones Anteriores

Escenario:

Tabla original:

| id | nombre |
| -- | ------ |
| 1  | Juan   |
| 2  | Ana    |

Posteriormente:

```sql id="6p8dxy"
DELETE FROM clientes
WHERE id = 2;
```

Tabla actual:

| id | nombre |
| -- | ------ |
| 1  | Juan   |

Con Time Travel es posible consultar la versión anterior donde Ana aún existía.

## Restauración de Bases de Datos

No solo se pueden recuperar tablas.

También es posible recuperar:

* Databases.
* Schemas.
* Tablas.
* Vistas.

Ejemplo:

```sql id="b3g6cu"
UNDROP DATABASE ventas_db;
```

Esto simplifica enormemente los procesos de recuperación.

## Beneficios de Time Travel

* Recuperación rápida.
* Menor dependencia de respaldos.
* Protección ante errores humanos.
* Auditoría histórica.
* Mayor continuidad operativa.

## Limitaciones de Time Travel

Aunque es una herramienta poderosa, no debe considerarse un reemplazo absoluto de una estrategia de respaldo.

Aspectos importantes:

* Tiene una ventana de retención limitada.
* Incrementa el almacenamiento utilizado.
* No conserva información indefinidamente.

## ¿Qué es Fail-Safe?

Fail-Safe es una capa adicional de protección administrada por Snowflake.

Su objetivo es proporcionar una recuperación de emergencia cuando el período de Time Travel ya ha expirado.

## Diferencia entre Time Travel y Fail-Safe

| Característica             | Time Travel | Fail-Safe |
| -------------------------- | ----------- | --------- |
| Acceso por usuario         | Sí          | No        |
| Recuperación inmediata     | Sí          | No        |
| Uso diario                 | Sí          | No        |
| Recuperación de emergencia | No          | Sí        |

## Flujo Conceptual

```text id="ozbffo"
Dato Actual
      ↓
Time Travel
      ↓
Fail-Safe
      ↓
Eliminación definitiva
```

Primero se utiliza Time Travel.

Si el período de Time Travel ya terminó, Snowflake aún conserva los datos temporalmente mediante Fail-Safe.

## Ejemplo de Recuperación

Situación:

```sql id="r54yzn"
DROP TABLE ventas;
```

Durante Time Travel:

```sql id="hqczh0"
UNDROP TABLE ventas;
```

Recuperación inmediata.

Si Time Travel expira:

```text id="gfdl7j"
Contacto con soporte Snowflake
```

El proceso de recuperación pasa a Fail-Safe.

## Consideraciones de Fail-Safe

* No está diseñado para operaciones diarias.
* Requiere intervención del soporte de Snowflake.
* Puede tomar más tiempo recuperar la información.
* Está orientado a situaciones críticas.

## Caso Empresarial

Una empresa financiera posee una tabla:

```text id="xyh9gs"
transacciones
```

Con:

```text id="1gc5cr"
500 millones de registros
```

Un administrador ejecuta accidentalmente:

```sql id="smqj7v"
DROP TABLE transacciones;
```

Gracias a Time Travel:

```sql id="zr4mrf"
UNDROP TABLE transacciones;
```

Resultado:

```text id="cz8a56"
Tabla recuperada
en pocos segundos
```

Sin necesidad de restaurar respaldos ni detener operaciones.

## Impacto en Gobernanza

Time Travel ofrece beneficios importantes para:

* Auditorías.
* Cumplimiento normativo.
* Investigación de incidentes.
* Recuperación ante errores operativos.

Por esta razón es una característica ampliamente utilizada en sectores como:

* Banca.
* Seguros.
* Telecomunicaciones.
* Salud.

## Buenas Prácticas

* Conocer el período de retención configurado.
* Utilizar Time Travel para recuperación operativa.
* Mantener respaldos externos para información crítica.
* Capacitar a los administradores sobre UNDROP.
* Documentar procedimientos de recuperación.
* No depender exclusivamente de Fail-Safe.

## Resumen

Time Travel y Fail-Safe son mecanismos de protección que permiten recuperar información eliminada o modificada accidentalmente en Snowflake. Time Travel proporciona acceso directo a versiones históricas de los datos y permite restauraciones rápidas, mientras que Fail-Safe ofrece una capa adicional de recuperación para situaciones excepcionales. Estas funcionalidades reducen significativamente el riesgo operativo y representan una de las ventajas más importantes de Snowflake frente a otras plataformas de datos modernas.


# Materialized Views y Caché en BigQuery y Snowflake

## Introducción

En ambientes analíticos es común que ciertas consultas se ejecuten cientos o miles de veces al día. Por ejemplo, dashboards ejecutivos, reportes de ventas o indicadores financieros.

Si cada consulta tuviera que recalcular los resultados desde cero, el consumo de recursos sería elevado y los tiempos de respuesta aumentarían considerablemente.

Para resolver este problema, tanto BigQuery como Snowflake incorporan mecanismos de optimización como Materialized Views y sistemas de caché.

## ¿Qué es una Materialized View?

Una Materialized View es una vista que almacena físicamente los resultados de una consulta.

A diferencia de una View tradicional, los datos ya se encuentran preprocesados, por lo que las consultas son mucho más rápidas.

## View Tradicional

```sql
CREATE VIEW v_ventas_producto AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas
GROUP BY producto;
```

Cada vez que se consulta:

```sql
SELECT *
FROM v_ventas_producto;
```

La consulta se recalcula desde la tabla original.

## Materialized View

```sql
CREATE MATERIALIZED VIEW mv_ventas_producto AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas
GROUP BY producto;
```

Consulta:

```sql
SELECT *
FROM mv_ventas_producto;
```

Resultado:

```text
Datos previamente calculados
```

Esto reduce significativamente los tiempos de ejecución.

## Materialized Views en BigQuery

BigQuery mantiene automáticamente las Materialized Views sincronizadas con la tabla origen.

Ejemplo:

Tabla:

```text
ventas
```

Materialized View:

```text
mv_ventas_producto
```

Cuando llegan nuevos registros:

```text
ventas
    ↓
Actualización automática
    ↓
mv_ventas_producto
```

El mantenimiento es administrado por BigQuery.

## Caso de Uso en BigQuery

Consulta frecuente:

```sql
SELECT
    producto,
    SUM(monto)
FROM ventas
GROUP BY producto;
```

Ejecutada por:

* Dashboard de Power BI.
* Dashboard de Looker.
* Reporte ejecutivo.

Sin Materialized View:

```text
Recalcular constantemente
```

Con Materialized View:

```text
Resultados preagregados
```

Beneficio:

* Menor costo.
* Menor tiempo de respuesta.

## Materialized Views en Snowflake

Snowflake ofrece una funcionalidad similar.

Ejemplo:

```sql
CREATE MATERIALIZED VIEW mv_ventas_producto AS
SELECT
    producto,
    SUM(monto) total_ventas
FROM ventas
GROUP BY producto;
```

La vista materializada se mantiene actualizada automáticamente.

Ventajas:

* Consultas más rápidas.
* Menor consumo de Warehouse.
* Menor tiempo de ejecución.

## ¿Cuándo Utilizar Materialized Views?

Escenarios recomendados:

* Consultas repetitivas.
* Dashboards.
* KPIs ejecutivos.
* Agregaciones complejas.
* Reportes diarios.

Ejemplos:

| Consulta            | Recomendado       |
| ------------------- | ----------------- |
| Ventas por mes      | Sí                |
| Ventas por producto | Sí                |
| Top clientes        | Sí                |
| Consulta ocasional  | No necesariamente |

## ¿Qué es el Caché?

El caché es un mecanismo que almacena temporalmente resultados previamente calculados.

Si la misma consulta se ejecuta nuevamente y los datos no han cambiado, el sistema puede devolver el resultado almacenado sin volver a procesar la información.

## Caché en BigQuery

Ejemplo:

Primera ejecución:

```sql
SELECT *
FROM clientes;
```

Tiempo:

```text
3 segundos
```

Segunda ejecución:

```sql
SELECT *
FROM clientes;
```

Tiempo:

```text
Menos de 1 segundo
```

BigQuery reutiliza el resultado almacenado.

## Beneficios del Caché en BigQuery

* Menor tiempo de respuesta.
* Menor procesamiento.
* Menor costo en ciertos escenarios.
* Mejor experiencia para los usuarios.

## Caché en Snowflake

Snowflake utiliza varios niveles de caché.

### Result Cache

Almacena resultados de consultas ejecutadas previamente.

Ejemplo:

```sql
SELECT *
FROM clientes;
```

Si la consulta se repite:

```text
Resultado recuperado desde caché
```

Sin necesidad de reprocesar los datos.

## Local Disk Cache

Los datos leídos recientemente permanecen temporalmente disponibles dentro del Warehouse.

Beneficios:

* Menor lectura desde almacenamiento.
* Mayor velocidad de respuesta.

## Metadata Cache

Snowflake también mantiene información estadística y metadatos que aceleran la optimización de consultas.

## Comparación General

| Característica            | BigQuery | Snowflake |
| ------------------------- | -------- | --------- |
| Materialized Views        | Sí       | Sí        |
| Caché de resultados       | Sí       | Sí        |
| Actualización automática  | Sí       | Sí        |
| Optimización transparente | Alta     | Alta      |
| Administración requerida  | Baja     | Baja      |

## Caso Empresarial

Una empresa posee un dashboard de ventas que se actualiza cada 10 minutos.

Consulta principal:

```sql
SELECT
    sucursal,
    SUM(monto)
FROM ventas
GROUP BY sucursal;
```

Registros:

```text
500 millones
```

Sin optimización:

```text
Recalcular constantemente
```

Con Materialized View:

```text
Resultados ya agregados
```

Beneficios:

* Dashboard más rápido.
* Menor costo.
* Menor carga de procesamiento.

## Cuándo NO Utilizar Materialized Views

No son necesarias cuando:

* La consulta se ejecuta ocasionalmente.
* Los datos cambian constantemente.
* La consulta es muy simple.
* El beneficio de rendimiento es mínimo.

En estos casos, una View tradicional suele ser suficiente.

## Buenas Prácticas

* Utilizar Materialized Views para consultas recurrentes.
* Monitorear el beneficio real obtenido.
* Evitar crear Materialized Views innecesarias.
* Aprovechar los mecanismos de caché nativos.
* Revisar periódicamente el consumo de almacenamiento adicional.

## Resumen

Las Materialized Views y los mecanismos de caché permiten acelerar significativamente las consultas tanto en BigQuery como en Snowflake. Las Materialized Views almacenan resultados previamente calculados, mientras que el caché reutiliza resultados y datos consultados recientemente. Utilizadas correctamente, estas herramientas reducen tiempos de respuesta, optimizan recursos y mejoran la experiencia de los usuarios en entornos analíticos empresariales.


# Integración con Airflow

## Introducción

En proyectos reales de Ingeniería de Datos, BigQuery y Snowflake rara vez funcionan de forma aislada. Normalmente forman parte de pipelines automatizados que extraen, transforman y cargan información desde múltiples fuentes.

Aquí es donde Apache Airflow juega un papel fundamental, permitiendo orquestar procesos de manera programada y controlada.

La combinación Airflow + BigQuery o Airflow + Snowflake es una de las arquitecturas más utilizadas actualmente en entornos empresariales.

## Arquitectura General

Flujo típico:

```text
ERP / CRM / APIs
        ↓
   Airflow
        ↓
 BigQuery / Snowflake
        ↓
 Power BI / Tableau / Looker
```

Airflow coordina cada una de las etapas del proceso.

## Integración con BigQuery

Airflow dispone de operadores específicos para interactuar con BigQuery.

Algunas tareas comunes:

* Ejecutar consultas SQL.
* Crear tablas.
* Cargar datos.
* Validar información.
* Ejecutar procesos ELT.

## Ejecución de Consultas

Ejemplo conceptual:

```python
from airflow.providers.google.cloud.operators.bigquery import BigQueryInsertJobOperator

consulta_ventas = BigQueryInsertJobOperator(
    task_id="ventas_diarias",
    configuration={
        "query": {
            "query": """
                SELECT
                    producto,
                    SUM(monto)
                FROM ventas
                GROUP BY producto
            """,
            "useLegacySql": False
        }
    }
)
```

Proceso:

```text
Airflow
    ↓
BigQuery
    ↓
Resultado
```

## Caso Práctico BigQuery

# Integración con dbt



## Introducción

En las arquitecturas modernas de datos, una de las combinaciones más utilizadas es:

```text
Airflow
   ↓
dbt
   ↓
BigQuery / Snowflake
```

Mientras Airflow se encarga de la orquestación, dbt (Data Build Tool) se especializa en las transformaciones de datos.

Actualmente, dbt es una de las herramientas más demandadas dentro de la Ingeniería de Datos debido a que permite implementar ELT de forma organizada, reutilizable y fácil de mantener.

## ¿Qué es dbt?

dbt es una herramienta que permite transformar datos utilizando SQL.

Su filosofía es simple:

```text
Extraer
   ↓
Cargar
   ↓
Transformar
```

Es decir:

* Los datos se cargan primero en BigQuery o Snowflake.
* Las transformaciones se ejecutan posteriormente utilizando SQL.

Este enfoque se conoce como ELT.

## Arquitectura General

Ejemplo:

```text
ERP
 ↓
Airflow
 ↓
BigQuery / Snowflake
 ↓
dbt
 ↓
Tablas Analíticas
 ↓
Power BI / Tableau
```

dbt trabaja directamente dentro del motor de datos.

## Componentes Principales de dbt

### Models

Son consultas SQL que generan tablas o vistas.

Ejemplo:

```sql
SELECT
    id_cliente,
    nombre,
    pais
FROM clientes
WHERE estado = 'ACTIVO'
```

Archivo:

```text
clientes_activos.sql
```

Cuando dbt ejecuta el modelo, crea una tabla o vista dentro de BigQuery o Snowflake.

## Sources

Representan las tablas de origen.

Ejemplo:

```yaml
sources:
  - name: ventas
    tables:
      - name: clientes
```

Beneficios:

* Documentación.
* Validación.
* Trazabilidad.

## Tests

Permiten validar la calidad de los datos.

Ejemplo:

```yaml
columns:
  - name: id_cliente
    tests:
      - unique
      - not_null
```

Validaciones:

```text
No duplicados
No valores nulos
```

## Documentation

dbt genera documentación automática.

Información generada:

* Tablas.
* Dependencias.
* Columnas.
* Relaciones.

Esto facilita enormemente el mantenimiento de proyectos grandes.

## Ejemplo de Flujo ELT

Tabla origen:

```text
clientes
```

Contenido:

| id_cliente | nombre | estado   |
| ---------- | ------ | -------- |
| 1          | Juan   | ACTIVO   |
| 2          | Ana    | INACTIVO |
| 3          | Carlos | ACTIVO   |

Modelo dbt:

```sql
SELECT
    id_cliente,
    nombre
FROM clientes
WHERE estado = 'ACTIVO'
```

Resultado:

| id_cliente | nombre |
| ---------- | ------ |
| 1          | Juan   |
| 3          | Carlos |

## Materializations

dbt permite definir cómo se almacenarán los resultados.

### View

```yaml
materialized: view
```

Resultado:

```text
Vista
```

### Table

```yaml
materialized: table
```

Resultado:

```text
Tabla física
```

### Incremental

```yaml
materialized: incremental
```

Resultado:

```text
Carga únicamente nuevos registros
```

Muy utilizado en producción.

## Modelos por Capas

Una práctica común consiste en organizar los modelos en capas.

### Staging

Datos limpiados mínimamente.

```text
stg_clientes
stg_ventas
```

### Intermediate

Transformaciones intermedias.

```text
int_clientes
int_ventas
```

### Mart

Tablas listas para análisis.

```text
dim_cliente
fact_ventas
```

Arquitectura:

```text
Raw
 ↓
Staging
 ↓
Intermediate
 ↓
Mart
```

## dbt con BigQuery

dbt genera consultas SQL que se ejecutan directamente en BigQuery.

Proceso:

```text
dbt
 ↓
SQL
 ↓
BigQuery
 ↓
Resultado
```

Ventajas:

* Escalabilidad.
* Serverless.
* Fácil implementación.

## dbt con Snowflake

Funcionamiento similar.

Proceso:

```text
dbt
 ↓
SQL
 ↓
Snowflake
 ↓
Resultado
```

Ventajas:

* Excelente rendimiento.
* Alta concurrencia.
* Integración nativa.

## Integración con Airflow

Proceso típico:

```text
Carga de Datos
        ↓
dbt run
        ↓
dbt test
        ↓
Actualización de Dashboards
```

Ejemplo conceptual:

```python
carga >> dbt_run >> dbt_test >> dashboard
```

Airflow controla la ejecución de dbt dentro del pipeline.

## Caso Empresarial

Una empresa de retail recibe diariamente:

* Ventas.
* Inventarios.
* Clientes.

Proceso:

```text
ERP
 ↓
Airflow
 ↓
BigQuery
 ↓
dbt
 ↓
fact_ventas
dim_cliente
dim_producto
 ↓
Power BI
```

Beneficios:

* Transformaciones centralizadas.
* Código reutilizable.
* Mayor calidad de datos.

## Ventajas de dbt

* Desarrollo basado en SQL.
* Control de versiones mediante Git.
* Documentación automática.
* Pruebas de calidad integradas.
* Arquitectura modular.
* Fácil mantenimiento.

## Buenas Prácticas

* Organizar modelos por capas.
* Utilizar pruebas automáticas.
* Documentar modelos críticos.
* Mantener nombres consistentes.
* Utilizar cargas incrementales cuando sea posible.
* Versionar todo el proyecto en Git.

## Arquitectura Moderna Recomendada

Actualmente una de las arquitecturas más utilizadas es:

```text
Cloud Storage
       ↓
Airflow
       ↓
BigQuery / Snowflake
       ↓
dbt
       ↓
Data Mart
       ↓
Power BI / Tableau
```

Esta arquitectura permite construir plataformas escalables, mantenibles y alineadas con las mejores prácticas de la industria.

## Resumen

dbt se ha convertido en una herramienta fundamental dentro de las arquitecturas modernas de Ingeniería de Datos. Su enfoque basado en SQL permite implementar transformaciones mediante el paradigma ELT, aprovechando la potencia de BigQuery y Snowflake. Combinado con Airflow, proporciona una solución robusta para construir pipelines escalables, mantenibles y orientados a la calidad de los datos.


# Monitoreo y Observabilidad

## Introducción

En cualquier plataforma de datos moderna no basta con cargar y transformar información. También es necesario monitorear continuamente el estado de los procesos, consultas, recursos y costos para detectar problemas antes de que impacten al negocio.

La observabilidad permite responder preguntas como:

* ¿Por qué falló un pipeline?
* ¿Qué consulta está consumiendo más recursos?
* ¿Quién ejecutó determinada operación?
* ¿Por qué aumentaron los costos este mes?
* ¿Qué proceso está generando cuellos de botella?

Tanto BigQuery como Snowflake proporcionan herramientas para monitorear el comportamiento de la plataforma y facilitar la administración diaria.

## ¿Qué Debe Monitorearse?

Las áreas más importantes son:

* Consultas SQL.
* Costos.
* Consumo de recursos.
* Procesos ETL y ELT.
* Accesos y seguridad.
* Calidad de datos.

## Monitoreo en BigQuery

BigQuery registra información detallada sobre cada consulta ejecutada.

Información disponible:

* Usuario.
* Fecha y hora.
* Tiempo de ejecución.
* Datos procesados.
* Estado de ejecución.

Ejemplo conceptual:

| Usuario | Datos Procesados | Duración |
| ------- | ---------------- | -------- |
| Juan    | 10 GB            | 5 seg    |
| Ana     | 2 TB             | 45 seg   |
| Carlos  | 500 MB           | 2 seg    |

Esta información permite identificar consultas costosas o ineficientes.

## Historial de Consultas

Vista utilizada frecuentemente:

```sql
SELECT *
FROM region-us.INFORMATION_SCHEMA.JOBS;
```

Información disponible:

* Query ejecutada.
* Usuario.
* Estado.
* Bytes procesados.
* Tiempo de ejecución.

## Monitoreo de Costos en BigQuery

Aspectos importantes:

* Datos procesados diariamente.
* Consultas más costosas.
* Usuarios con mayor consumo.
* Crecimiento del almacenamiento.

Indicadores comunes:

| Métrica            | Objetivo              |
| ------------------ | --------------------- |
| TB procesados      | Controlar crecimiento |
| Consultas fallidas | Reducir errores       |
| Costos mensuales   | Mantener presupuesto  |

## Monitoreo en Snowflake

Snowflake proporciona múltiples vistas administrativas para supervisar la actividad de la plataforma.

Información disponible:

* Consultas ejecutadas.
* Créditos consumidos.
* Uso de Warehouses.
* Usuarios conectados.
* Historial de cargas.

## Historial de Consultas

Consulta conceptual:

```sql
SELECT *
FROM SNOWFLAKE.ACCOUNT_USAGE.QUERY_HISTORY;
```

Información disponible:

* Usuario.
* Consulta.
* Duración.
* Warehouse utilizado.
* Recursos consumidos.

## Monitoreo de Warehouses

Aspectos importantes:

* Tiempo activo.
* Consumo de créditos.
* Concurrencia.
* Utilización.

Ejemplo:

| Warehouse | Estado     |
| --------- | ---------- |
| BI_WH     | Activo     |
| ETL_WH    | Suspendido |
| DS_WH     | Activo     |

Esto permite detectar recursos sobredimensionados o infrautilizados.

## Monitoreo de Pipelines

Además de BigQuery y Snowflake, los pipelines también deben supervisarse.

Ejemplo con Airflow:

| DAG            | Estado  |
| -------------- | ------- |
| carga_clientes | Success |
| carga_ventas   | Success |
| resumen_diario | Failed  |

Información relevante:

* Duración.
* Errores.
* Reintentos.
* Dependencias.

## Monitoreo de Calidad de Datos

Algunos indicadores importantes:

### Valores Nulos

Ejemplo:

| Tabla    | % Nulos |
| -------- | ------- |
| clientes | 0%      |
| ventas   | 2%      |

### Registros Duplicados

Ejemplo:

| Tabla    | Duplicados |
| -------- | ---------- |
| clientes | 0          |

Ejemplo:

| Fecha | Registros |
| ----- | --------- |
| 14/06 | 100,000   |
| 15/06 | 15,000    |

Una caída repentina puede indicar problemas en la carga.

Arquitectura típica:

## Resumen

El monitoreo y la observabilidad son componentes esenciales de cualquier plataforma moderna de datos. Tanto BigQuery como Snowflake proporcionan mecanismos para supervisar consultas, costos, recursos y actividad de usuarios. Combinados con herramientas como Airflow y dbt, permiten construir entornos más confiables, escalables y fáciles de administrar en producción.


# Buenas Prácticas para Producción

## Introducción

La diferencia entre un entorno de pruebas y un entorno productivo no suele estar en la tecnología utilizada, sino en la forma en que se implementa y administra.

Una solución puede funcionar correctamente durante las pruebas, pero presentar problemas de rendimiento, costos, seguridad o mantenimiento cuando comienza a operar con grandes volúmenes de datos y múltiples usuarios.

Por esta razón, es fundamental aplicar buenas prácticas desde el diseño inicial de la plataforma.

## Diseñar Pensando en la Escalabilidad

Una de las reglas más importantes es asumir que los datos crecerán con el tiempo.

Diseño incorrecto:

```text
10,000 registros actuales
```

Diseño recomendado:

```text
10,000 registros actuales
↓
10 millones futuros
↓
100 millones futuros
```

La arquitectura debe soportar este crecimiento sin requerir rediseños constantes.

## Separar Ambientes

Nunca se debe trabajar directamente en producción.

Estructura recomendada:

```text
DEV
 ↓
QA
 ↓
PROD
```

Ejemplo:

```text
empresa-dev
empresa-qa
empresa-prod
```

Beneficios:

* Menor riesgo.
* Validación previa.
* Mayor control de cambios.

## Implementar Control de Versiones

Todo código debe almacenarse en Git.

Elementos a versionar:

* DAGs de Airflow.
* Modelos dbt.
* Scripts SQL.
* Configuraciones.
* Infraestructura como código.

Beneficios:

* Historial de cambios.
* Trabajo colaborativo.
* Recuperación de versiones anteriores.

## Aplicar el Principio de Mínimo Privilegio

Los usuarios deben tener únicamente los permisos necesarios.

Incorrecto:

```text
Todos son administradores
```

Correcto:

```text
Analista → Lectura
Data Engineer → Escritura
Administrador → Gestión
```

Esto reduce significativamente los riesgos operativos.

## Automatizar Todo lo Posible

Procesos manuales generan errores.

Incorrecto:

```text
Descargar archivo
↓
Subir archivo
↓
Ejecutar consulta manualmente
```

Correcto:

```text
Airflow
↓
Carga automática
↓
dbt
↓
Dashboard
```

Beneficios:

* Menos errores.
* Mayor consistencia.
* Mayor productividad.

## Utilizar ELT Moderno

Arquitectura recomendada:

```text
Fuente
 ↓
BigQuery / Snowflake
 ↓
dbt
 ↓
Data Mart
```

Ventajas:

* Menor complejidad.
* Mayor rendimiento.
* Transformaciones centralizadas.

## Optimizar Costos Permanentemente

### BigQuery

* Evitar SELECT *.
* Utilizar particiones.
* Utilizar clustering.
* Crear Materialized Views.

### Snowflake

* Configurar Auto Suspend.
* Configurar Auto Resume.
* Monitorear créditos.
* Ajustar tamaños de Warehouse.

La optimización debe ser continua.

## Monitorear la Plataforma

Aspectos mínimos a supervisar:

* Costos.
* Consultas lentas.
* Pipelines fallidos.
* Consumo de recursos.
* Calidad de datos.

No esperar a que los usuarios reporten problemas.

## Implementar Validaciones de Calidad

Ejemplos:

### Valores nulos

```text
id_cliente = NULL
```

### Duplicados

```text
cliente repetido
```

### Fechas inválidas

```text
31/02/2026
```

Estas validaciones deben ejecutarse automáticamente.

## Documentar los Procesos

Toda plataforma debe responder fácilmente:

* ¿De dónde provienen los datos?
* ¿Quién mantiene el proceso?
* ¿Qué transforma cada tabla?
* ¿Qué dependencias existen?

Herramientas como dbt ayudan significativamente en esta tarea.

## Diseñar para la Recuperación

Siempre debe existir un plan ante fallos.

Ejemplos:

* Recuperación mediante Time Travel.
* Restauración de respaldos.
* Reejecución de pipelines.
* Procedimientos de contingencia.

La recuperación debe estar documentada y probada.

## Organizar los Datos por Dominios

Ejemplo:

```text
ventas
marketing
rrhh
finanzas
```

Evitar:

```text
dataset1
dataset2
dataset3
```

Los nombres deben reflejar claramente el propósito de los datos.

## Utilizar Nombres Consistentes

Ejemplo recomendado:

```text
stg_clientes
stg_ventas

dim_cliente
dim_producto

fact_ventas
```

Beneficios:

* Fácil mantenimiento.
* Mayor comprensión.
* Menor curva de aprendizaje.

## Mantener Arquitecturas Simples

Error común:

```text
20 herramientas
15 integraciones
10 capas innecesarias
```

Buena práctica:

```text
Airflow
 ↓
BigQuery / Snowflake
 ↓
dbt
 ↓
Power BI
```

La simplicidad suele ser más sostenible.

## Arquitectura Recomendada

Para la mayoría de organizaciones:

```text
Fuentes de Datos
        ↓
Airflow
        ↓
Cloud Storage
        ↓
BigQuery / Snowflake
        ↓
dbt
        ↓
Data Mart
        ↓
Power BI / Tableau
```

Esta arquitectura cubre la mayoría de necesidades empresariales modernas.

## Checklist de Producción

Antes de liberar un proyecto verificar:

### Seguridad

* Roles configurados.
* Permisos revisados.
* Accesos auditados.

### Costos

* Particiones implementadas.
* Warehouses optimizados.
* Consultas revisadas.

### Calidad

* Validaciones implementadas.
* Datos consistentes.
* Pruebas ejecutadas.

### Operación

* Monitoreo activo.
* Alertas configuradas.
* Documentación disponible.

### Recuperación

* Procedimientos definidos.
* Time Travel validado.
* Estrategias de respaldo disponibles.

## Errores Más Comunes

### No controlar costos

Especialmente frecuente en BigQuery.

### Mantener Warehouses activos

Especialmente frecuente en Snowflake.

### Exceso de permisos

Genera riesgos de seguridad.

### Falta de monitoreo

Los problemas se detectan demasiado tarde.

### Procesos manuales

Incrementan errores y reducen escalabilidad.

## Resumen

Las buenas prácticas para producción buscan garantizar que una plataforma de datos sea escalable, segura, eficiente y fácil de mantener. La combinación de Airflow, BigQuery o Snowflake, dbt, monitoreo continuo, control de costos y una adecuada gobernanza permite construir soluciones robustas alineadas con las necesidades actuales de la Ingeniería de Datos. Más allá de la tecnología utilizada, el éxito de una plataforma depende de la disciplina operativa con la que se implementan estas prácticas.

# Parte 8 - Consultas Analíticas

## Top 10 Productos

```sql
SELECT
    p.producto,
    SUM(v.monto) total_ventas
FROM retail_dw.ventas v
JOIN retail_dw.productos p
ON v.id_producto = p.id_producto
GROUP BY p.producto
ORDER BY total_ventas DESC
LIMIT 10;
```

---

## Ventas por País

```sql
SELECT
    c.pais,
    SUM(v.monto) total_ventas
FROM retail_dw.ventas v
JOIN retail_dw.clientes c
ON v.id_cliente = c.id_cliente
GROUP BY c.pais;
```

---

## Ventas Mensuales

```sql
SELECT
    EXTRACT(MONTH FROM fecha_venta) mes,
    SUM(monto) ventas
FROM retail_dw.ventas
GROUP BY mes;
```

# Parte 9 - Optimización

## Crear Tabla Particionada

```sql
CREATE TABLE retail_dw.ventas_particionadas
(
    id_venta INT64,
    fecha_venta DATE,
    id_cliente INT64,
    id_producto INT64,
    sucursal STRING,
    monto NUMERIC
)
PARTITION BY fecha_venta;
```

---

## Crear Clustering

```sql
CREATE TABLE retail_dw.ventas_cluster
(
    id_venta INT64,
    fecha_venta DATE,
    id_cliente INT64,
    id_producto INT64,
    sucursal STRING,
    monto NUMERIC
)
PARTITION BY fecha_venta
CLUSTER BY sucursal;
```

# Parte 10 - Materialized View

Crear:

```sql
CREATE MATERIALIZED VIEW retail_dw.mv_ventas_producto AS
SELECT
    id_producto,
    SUM(monto) total_ventas
FROM retail_dw.ventas
GROUP BY id_producto;
```
