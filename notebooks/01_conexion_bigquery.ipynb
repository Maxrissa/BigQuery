# Conexión desde Python

```python
from google.cloud import bigquery
import os

print("Conectando a BigQuery")

# 1. Verificar que el archivo JSON existe
ruta_json = "../credentials/service_account.json"

if not os.path.exists(ruta_json):
    print(f"No se encontró el archivo: {ruta_json}")
    print("   Asegúrate de tener service_account.json en la carpeta credentials/")
else:
    print(f"Archivo de credenciales encontrado")
    
    # 2. Conectar a BigQuery
    try:
        client = bigquery.Client.from_service_account_json(
            ruta_json,
            project='project-6daabb75-e226-428c-869'
        )
        print("Conexión exitosa a BigQuery")
        print(f"   Proyecto: {client.project}")
    except Exception as e:
        print(f" Error de conexión: {e}")
```
