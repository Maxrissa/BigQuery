## Importaciones

# Generacion de clientes

```python
def generar_clientes():
    clientes = []

    clientes_por_pais = TOTAL_CLIENTES // len(PAISES)

    id_cliente = 1

    for pais in PAISES:
        for _ in range(clientes_por_pais):
            clientes.append({
                "id_cliente": id_cliente,
                "nombre": fake.name(),
                "pais": pais
            })

            id_cliente += 1

    df_clientes = pd.DataFrame(clientes)

    return df_clientes
```

```python
df_clientes = generar_clientes()
```

```python
print("Cantidad total de clientes:", len(df_clientes))

print("\nClientes por pais:")
print(df_clientes["pais"].value_counts())
```

# Generación de productos

```python
def generar_productos():
    productos = []

    for id_producto in range(1, TOTAL_PRODUCTOS + 1):
        categoria = random.choice(list(PRODUCTOS_POR_CATEGORIA.keys()))
        nombre_base = random.choice(PRODUCTOS_POR_CATEGORIA[categoria])

        productos.append({
            "id_producto": id_producto,
            "producto": f"{nombre_base}",
            "categoria": categoria
        })

    return pd.DataFrame(productos)
```

```python
df_productos = generar_productos()
```

## Validamos los datos 

```python
print("Cantidad total de productos:", len(df_productos))

print("\nProductos por categoria:")
print(df_productos["categoria"].value_counts())
```

# Generación de ventas

```python
def generar_ventas():
    ventas = []

    fecha_inicio = datetime(2026, 1, 1)
    fecha_fin = datetime(2026, 12, 31)

    dias_totales = (fecha_fin - fecha_inicio).days

    for id_venta in range(1, TOTAL_VENTAS + 1):

        fecha_venta = fecha_inicio + timedelta(
            days=random.randint(0, dias_totales)
        )

        ventas.append({
            "id_venta": id_venta,
            "fecha_venta": fecha_venta.date(),
            "id_cliente": random.randint(1, TOTAL_CLIENTES),
            "id_producto": random.randint(1, TOTAL_PRODUCTOS),
            "sucursal": random.choice(SUCURSALES),
            "monto": round(random.uniform(50, 1000), 2)
        })

    return pd.DataFrame(ventas)
```

```python
df_ventas = generar_ventas()
```

```python
df_ventas.head()
```

## Validamos los datos 

```python
print("TOTAL VENTAS:", len(df_ventas))

print("\nFECHAS:")
print(df_ventas["fecha_venta"].min())
print(df_ventas["fecha_venta"].max())

print("\nVENTAS POR SUCURSAL:")
print(df_ventas["sucursal"].value_counts())
```
